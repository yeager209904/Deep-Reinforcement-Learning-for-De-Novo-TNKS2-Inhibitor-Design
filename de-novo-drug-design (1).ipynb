```python
#!/usr/bin/env python3
"""
Reinforcement Learning for Standalone Molecular Generation
Fine-tunes a model to optimize for specific properties.
"""
import pickle
import torch
import torch.optim as optim
import numpy as np
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Descriptors, Crippen
from rdkit.DataStructs import TanimotoSimilarity, BulkTanimotoSimilarity
from rdkit.Chem.rdMolDescriptors import GetMorganFingerprintAsBitVect

# Import classes from your base training script
from molecule_generator import MolecularLSTM, SMILESDataset, MolecularTrainer, SMILESVocabulary

class ReinforcementLearningTrainer(MolecularTrainer):
    """
    Extends the trainer for Reinforcement Learning using a custom scoring function.
    """
    def __init__(self, model, vocab, agent_model_path, device='cuda' if torch.cuda.is_available() else 'cpu'):
        super().__init__(model, vocab, device)
        
        # Load the pretrained agent model from transfer learning
        if os.path.exists(agent_model_path):
            print(f"Loading agent model from {agent_model_path}")
            self.model.load_state_dict(torch.load(agent_model_path, map_location=device))
            print("Agent model loaded successfully!")
        else:
            raise FileNotFoundError(f"Agent model not found at {agent_model_path}")

    def score_molecules(self, smiles_list, target_smiles_list):
        """
        Calculates a score for each molecule based on multiple objectives.
        This is your RL 'Environment'.
        """
        scores = []
        target_fps = [GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(s), 2, nBits=2048) for s in target_smiles_list]

        for smiles in smiles_list:
            mol = Chem.MolFromSmiles(smiles)
            if mol is None:
                scores.append(0.0) # Penalize invalid molecules
                continue
            
            # --- Objective 1: Similarity to target (TNKS2 inhibitors) ---
            mol_fp = GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
            similarity = max(BulkTanimotoSimilarity(mol_fp, target_fps))
            sim_score = similarity # Score is the similarity itself

            # --- Objective 2: QED (Drug-likeness) ---
            qed_score = Descriptors.qed(mol)

            # --- Objective 3: Molecular Weight ---
            mw = Descriptors.MolWt(mol)
            mw_score = 1.0 if 150 < mw < 500 else 0.0 # Simple step function

            # --- Combine scores (weighted sum) ---
            # Adjust weights based on your priorities
            total_score = (0.5 * sim_score) + (0.3 * qed_score) + (0.2 * mw_score)
            scores.append(total_score)
            
        return np.array(scores)

    def train_reinforcement_learning(self, target_smiles, num_steps=5000, batch_size=128, lr=0.0001):
        """
        The main RL training loop (uses a simplified policy gradient approach).
        """
        optimizer = optim.Adam(self.model.parameters(), lr=lr)
        self.history = {'step': [], 'avg_score': [], 'validity': []}
        
        print(f"Starting Reinforcement Learning for {num_steps} steps...")

        for step in tqdm(range(num_steps), desc="RL Steps"):
            # 1. Sample a batch of molecules from the current agent
            sampled_smiles = self.sample_molecules(num_samples=batch_size, max_length=100)
            valid_smiles = [s for s in sampled_smiles if Chem.MolFromSmiles(s)]
            
            if not valid_smiles:
                continue

            # 2. Score the batch of molecules (get reward from environment)
            rewards = self.score_molecules(valid_smiles, target_smiles)

            # 3. Policy Gradient Update (simplified REINFORCE algorithm)

            # --------------------- THE FIX IS HERE ---------------------
            self.model.train()  # Switch model back to training mode before learning
            # ---------------------------------------------------------
            
            optimizer.zero_grad()
            
            # Re-calculate the log-probabilities of the actions (tokens) that led to these molecules
            encoded_smiles = [self.vocab.encode(s) for s in valid_smiles]
            
            total_loss = 0
            for i, enc in enumerate(encoded_smiles):
                input_seq = torch.tensor(enc[:-1], dtype=torch.long).unsqueeze(0).to(self.device)
                target_seq = torch.tensor(enc[1:], dtype=torch.long).unsqueeze(0).to(self.device)

                output, _ = self.model(input_seq)
                log_probs = torch.log_softmax(output, dim=2)
                
                # Select the log-probs of the tokens that were actually generated
                action_log_probs = log_probs.gather(2, target_seq.unsqueeze(2)).squeeze(-1)
                
                # Calculate loss: -log_prob * reward
                # We want to maximize the log-prob of good actions (high reward)
                loss = - (action_log_probs.mean() * rewards[i])
                total_loss += loss

            # Backpropagate the total loss for the batch
            (total_loss / len(valid_smiles)).backward()
            optimizer.step()

            # --- Logging ---
            if step % 50 == 0:
                avg_score = np.mean(rewards)
                validity = (len(valid_smiles) / batch_size) * 100
                self.history['step'].append(step)
                self.history['avg_score'].append(avg_score)
                self.history['validity'].append(validity)
                print(f"Step {step}/{num_steps} | Avg Score: {avg_score:.3f} | Validity: {validity:.1f}%")

        print("Reinforcement learning completed!")
        return self.history


def main():
    """Main function to run the RL experiment"""
    # --- PREREQUISITES ---
    # Ensure you have the vocabulary and the fine-tuned agent model from previous steps
    if not os.path.exists('models/vocab.pkl') or not os.path.exists('models/best_transfer_model.pt'):
        print("Error: vocab.pkl or best_transfer_model.pt not found. Run previous scripts first.")
        return

    # --- SETUP ---
    # Load vocabulary
    with open('models/vocab.pkl', 'rb') as f:
        vocab = pickle.load(f)

    # Initialize model architecture
    model = MolecularLSTM(vocab_size=vocab.vocab_size, embedding_dim=256, hidden_dim=512, num_layers=3, dropout=0.2)
    
    # Load your TNKS2 target SMILES from your csv file
    # This guides the similarity search
    import pandas as pd
    tnks2_df = pd.read_csv('data/tnks2.csv') # Make sure this path is correct
    target_smiles_list = tnks2_df.sort_values(by='exp (nM)').head(10)['SMILES'].tolist()
    print(f"Loaded {len(target_smiles_list)} target SMILES for similarity scoring.")

    # --- EXECUTION ---
    # Initialize RL trainer
    trainer = ReinforcementLearningTrainer(
        model=model,
        vocab=vocab,
        agent_model_path='models/best_transfer_model.pt'
    )
    
    # Run reinforcement learning
    history = trainer.train_reinforcement_learning(
        target_smiles=target_smiles_list,
        num_steps=2000, # Start with a smaller number of steps
        batch_size=64,
        lr=5e-5 # Use a very small learning rate for RL
    )

    # --- RESULTS ---
    # Save the final, RL-optimized model
    torch.save(model.state_dict(), 'models/final_rl_model.pt')
    print("Saved final RL-optimized model to models/final_rl_model.pt")

    # Generate and save a large batch of final molecules
    final_molecules = trainer.sample_molecules(num_samples=5000)
    with open('outputs/rl_generated_molecules.smi', 'w') as f:
        for smiles in final_molecules:
            f.write(f"{smiles}\n")
    print("Saved 5,000 generated molecules to outputs/rl_generated_molecules.smi")

    # Plot results
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['step'], history['avg_score'], label='Average Score')
    plt.xlabel('Step')
    plt.ylabel('Score')
    plt.title('RL Average Score Over Time')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['step'], history['validity'], label='Validity (%)', color='orange')
    plt.xlabel('Step')
    plt.ylabel('Validity (%)')
    plt.title('Sample Validity During RL')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('outputs/rl_training_history.png')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
# Path to your file (adjust if different)
smi_path = r"C:\Users\aneru\OneDrive\Desktop\AMRITA\SEM-7\Reinforcement Learning\Endsem\REINVENT4-main\tryingsmth\outputs\rl_generated_molecules.smi"

# --- helper: load smiles from a .smi file (handles one SMILES per line) ---
def load_smiles_from_smi(path):
    smiles = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            s = line.strip()
            if not s:
                continue
            # .smi files sometimes contain "SMILES <whitespace> name" -> take only first token
            smiles.append(s.split()[0])
    return smiles

# --- Try to use evaluate_generated_set if already present, else define minimal required functions ---
try:
    evaluate_generated_set  # if defined, do nothing
except NameError:
    # Minimal re-definition of core utilities (a reduced self-contained version).
    from rdkit import Chem
    from rdkit.Chem import rdMolDescriptors, Crippen, AllChem
    from rdkit.Chem.Scaffolds import MurckoScaffold
    import numpy as np

    def sanitize_and_canonicalize(smiles: str):
        try:
            m = Chem.MolFromSmiles(smiles)
            if m is None:
                return None
            return Chem.MolToSmiles(m, isomericSmiles=True)
        except Exception:
            return None

    def compute_validity_and_canonical(generated_smiles):
        canon = [sanitize_and_canonicalize(s) for s in generated_smiles]
        valid = [c is not None for c in canon]
        return valid, canon

    def uniqueness_fraction(canonical_smiles):
        valid_canon = [s for s in canonical_smiles if s is not None]
        if len(valid_canon) == 0:
            return 0.0
        return len(set(valid_canon)) / len(valid_canon)

    def average_tanimoto_diversity(canonical_smiles, fp_radius=2):
        valid_canon = [s for s in canonical_smiles if s is not None]
        n = len(valid_canon)
        if n < 2:
            return 0.0
        fps = []
        for s in valid_canon:
            m = Chem.MolFromSmiles(s)
            fps.append(AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=2048))
        sims = []
        import rdkit.DataStructs as DS
        for i in range(n):
            for j in range(i+1, n):
                sims.append(DS.TanimotoSimilarity(fps[i], fps[j]))
        sims = np.array(sims)
        return float(1.0 - sims.mean()) if sims.size > 0 else 0.0

    def scaffold_diversity_fraction(canonical_smiles):
        valid_canon = [s for s in canonical_smiles if s is not None]
        if len(valid_canon) == 0:
            return 0.0
        scaffolds = set()
        for s in valid_canon:
            m = Chem.MolFromSmiles(s)
            try:
                scaf = MurckoScaffold.GetScaffoldForMol(m)
                scaf_smiles = Chem.MolToSmiles(scaf, isomericSmiles=True) if scaf is not None else ""
            except Exception:
                scaf_smiles = ""
            scaffolds.add(scaf_smiles)
        return len(scaffolds) / len(valid_canon)

    def property_distributions(canonical_smiles):
        valid_canon = [s for s in canonical_smiles if s is not None]
        out = {"MolWt": {}, "LogP": {}, "TPSA": {}}
        if len(valid_canon) == 0:
            for k in out:
                out[k] = {"mean": float('nan'), "median": float('nan'), "std": float('nan')}
            return out
        molw, logp, tpsa = [], [], []
        for s in valid_canon:
            m = Chem.MolFromSmiles(s)
            molw.append(rdMolDescriptors.CalcExactMolWt(m))
            logp.append(Crippen.MolLogP(m))
            tpsa.append(rdMolDescriptors.CalcTPSA(m))
        def stats(arr):
            arr = np.array(arr)
            return {"mean": float(arr.mean()), "median": float(np.median(arr)), "std": float(arr.std())}
        out["MolWt"] = stats(molw)
        out["LogP"] = stats(logp)
        out["TPSA"] = stats(tpsa)
        return out

    def evaluate_generated_set(generated_smiles, reference_smiles=None, scores=None):
        valid_flags, canonical = compute_validity_and_canonical(generated_smiles)
        n = len(generated_smiles)
        n_valid = sum(valid_flags)
        frac_valid = n_valid / n if n > 0 else 0.0
        uniq = uniqueness_fraction(canonical)
        novel = float('nan')
        if reference_smiles is not None:
            ref_set = set(sanitize_and_canonicalize(s) for s in reference_smiles)
            valid_canon = [s for s in canonical if s is not None]
            novel = sum(1 for s in valid_canon if s not in ref_set) / len(valid_canon) if len(valid_canon) else 0.0
        diversity = average_tanimoto_diversity(canonical)
        scaffold_div = scaffold_diversity_fraction(canonical)
        props = property_distributions(canonical)
        return {
            "n_generated": n,
            "n_valid": n_valid,
            "validity_fraction": frac_valid,
            "uniqueness_fraction": uniq,
            "novelty_fraction": novel,
            "avg_tanimoto_distance": diversity,
            "scaffold_diversity_fraction": scaffold_div,
            "property_stats": props
        }

# --- Load RL-generated smiles ---
generated_smiles = load_smiles_from_smi(smi_path)
print(f"Loaded {len(generated_smiles)} SMILES from {smi_path}")

# If you have a training/reference .smi file, point to it to compute novelty.
# e.g. reference_path = "/mnt/data/training_smiles.smi"
reference_path = None
reference_smiles = None
if reference_path:
    reference_smiles = load_smiles_from_smi(reference_path)

# Run evaluation
metrics = evaluate_generated_set(generated_smiles, reference_smiles=reference_smiles, scores=None)

# Print a short summary
print("=== RL Generated Set Metrics ===")
print(f"Generated: {metrics['n_generated']}")
print(f"Valid: {metrics['n_valid']} ({metrics['validity_fraction']*100:.2f} %)")
print(f"Uniqueness: {metrics['uniqueness_fraction']*100:.2f} %")
if not np.isnan(metrics['novelty_fraction']):
    print(f"Novelty: {metrics['novelty_fraction']*100:.2f} %")
print(f"Avg Tanimoto distance (diversity): {metrics['avg_tanimoto_distance']:.4f}")
print(f"Scaffold diversity fraction: {metrics['scaffold_diversity_fraction']:.4f}")
print("Property stats (MolWt / LogP / TPSA):")
import json
print(json.dumps(metrics["property_stats"], indent=2))

# Save metrics JSON

out_json = r"C:\Users\aneru\OneDrive\Desktop\AMRITA\SEM-7\Reinforcement Learning\Endsem\REINVENT4-main\tryingsmth\outputs\rl_generated_molecules.json"
with open(out_json, "w", encoding="utf-8") as f:
    json.dump(metrics, f, indent=2)
print(f"Saved full metrics to {out_json}")
```

```python
"""
REINFORCE fine-tuning for MolGPT (SELFIES). Uses classes from your uploaded molecule_generator.py.
Save as molgpt_rl.py in the same folder as molecule_generator.py and run (on Kaggle).
This version updates paths for Kaggle:
 - Adds /kaggle/input/molecule-generator to sys.path so molecule_generator.py can be imported
 - Uses /kaggle/input/chembl-smi/chembl_molecules.smi as the default training SMILES file
 - Writes run checkpoints and outputs to /kaggle/working/runs/rl_molgpt
"""

import os
import sys
import json
import math
import random
from pathlib import Path
from typing import List, Tuple, Optional

import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from rdkit import Chem
from rdkit.Chem import QED
from rdkit.Chem import AllChem
import numpy as np
import selfies as sf

# Add Kaggle dataset folder that contains your molecule_generator.py (update if your dataset folder name differs)
# From the screenshot the folder name is "molecule-generator"
MOLECULE_GENERATOR_INPUT = "/kaggle/input/molecule-generator"
if MOLECULE_GENERATOR_INPUT not in sys.path:
    sys.path.append(MOLECULE_GENERATOR_INPUT)

# import model + tokenizer builder from uploaded file
# The file you uploaded contains MolGPTSmall and SELFIESTokenizer definitions; adapt import if filename differs
from molecule_generator import MolGPTSmall, SELFIESTokenizer, read_smiles, filter_and_canon, canon_smiles, smiles_to_selfies

# -------------------------
# Sampling with log-probs
# -------------------------
def sample_with_logprobs(model: nn.Module, tok: SELFIESTokenizer, device: str,
                         n_samples: int = 64, max_len: int = 256,
                         top_p: float = 0.95, temp: float = 1.0) -> Tuple[List[str], List[torch.Tensor]]:
    """
    Same behaviour but KEEP grad so log-probs have grad_fn.
    IMPORTANT: call this while model.train() (or at least not in a no_grad context).
    """
    # Do NOT wrap with torch.no_grad() here
    smiles_out = []
    logp_out = []
    for _ in range(n_samples):
        idx = torch.full((1,1), tok.bos, dtype=torch.long, device=device)
        log_probs = []
        for _step in range(max_len-1):
            logits, _ = model(idx)   # logits depends on model params (requires grad)
            logits = logits[:,-1,:] / float(max(1e-8, temp))
            probs = torch.softmax(logits, dim=-1)

            # nucleus filtering (keep as probabilities that depend on logits)
            sp, si = torch.sort(probs, descending=True)
            cum = torch.cumsum(sp, dim=-1)
            cutoff = cum > top_p
            cutoff[...,1:] = cutoff[...,:-1].clone()
            cutoff[...,0] = False
            sp = sp.clone()
            sp[cutoff] = 0.0
            sp = sp / sp.sum(dim=-1, keepdim=True)

            # sample from the filtered distribution
            dist = torch.distributions.Categorical(sp)
            next_token = dist.sample()                 # sampled index (tensor)
            next_logp = dist.log_prob(next_token)      # log prob (this has grad wrt logits via sp)
            idx = torch.cat([idx, next_token.unsqueeze(-1)], dim=1)
            log_probs.append(next_logp[0].unsqueeze(0))  # store scalar tensor (keeps grad)

            if next_token.item() == tok.eos:
                break

        # decode SELFIES -> SMILES (decoding itself is not part of graph)
        seq_ids = idx[0].tolist()
        se = tok.decode(seq_ids)
        smi = ""
        try:
            smi = canon_smiles(sf.decoder(se))
        except Exception:
            smi = ""
        smiles_out.append(smi if smi else "")
        if len(log_probs) == 0:
            # keep a tiny tensor so dimensions consistent; it won't contribute to grad
            logp_out.append(torch.tensor([0.0], device=device))
        else:
            logp_out.append(torch.cat(log_probs, dim=0))  # shape (T,) - has grad_fn
    return smiles_out, logp_out

# -------------------------
# Reward function
# -------------------------
def reward_function(smiles: str, reward_config: dict = None) -> float:
    """
    Compute scalar reward for a single canonical SMILES.
    reward_config (optional) keys:
      - validity_bonus: float added if valid (default 1.0)
      - qed_weight: multiply QED score (default 2.0)
      - penalize_large_mol: bool
      - max_atoms: int (default 60)
      - logp_weight: float (default 0.0) -- disabled unless you want to include
    """
    if reward_config is None:
        reward_config = {}
    validity_bonus = reward_config.get("validity_bonus", 1.0)
    qed_weight = reward_config.get("qed_weight", 2.0)
    penalize_large = reward_config.get("penalize_large_mol", True)
    max_atoms = reward_config.get("max_atoms", 60)

    if smiles is None or smiles == "":
        return 0.0
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return 0.0
    except Exception:
        return 0.0

    r = 0.0
    # validity
    r += validity_bonus
    # qed
    try:
        q = QED.qed(mol)
        r += qed_weight * float(q)
    except Exception:
        pass
    # penalize too large molecules (optional)
    if penalize_large and mol.GetNumAtoms() > max_atoms:
        r *= 0.5
    return float(r)

# -------------------------
# Reinforce training loop
# -------------------------
def reinforce_train(run_dir: str,
                    train_smi: Optional[str] = None,
                    ckpt_path: Optional[str] = None,
                    epochs: int = 200,
                    batch_size: int = 32,
                    lr: float = 2e-5,
                    max_len: int = 256,
                    top_p: float = 0.95,
                    temp: float = 1.0,
                    reward_config: dict = None,
                    device: Optional[str] = None,
                    eval_every: int = 5,
                    eval_n: int = 200):
    """
    Main REINFORCE fine-tuning loop.
      - run_dir: where to save vocab, checkpoints, logs
      - train_smi: path to dataset used to build vocab (MolGPT uses it to build tokenizer). If None, you must already have vocab.json in run_dir.
    """
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    os.makedirs(run_dir, exist_ok=True)

    # load or build vocab and tokenizer
    vocab_path = os.path.join(run_dir, "vocab.json")
    if train_smi and not os.path.exists(vocab_path):
        smiles = filter_and_canon(read_smiles(train_smi))
        toks = set()
        for s in smiles:
            se = smiles_to_selfies(s)
            if se:
                toks.update(sf.split_selfies(se))
        vocab = ["[PAD]","[BOS]","[EOS]"] + sorted(toks)
        json.dump(vocab, open(vocab_path, "w"))
    else:
        if not os.path.exists(vocab_path):
            raise FileNotFoundError(f"vocab.json not found in run_dir ({run_dir}). Provide train_smi to build vocab or place vocab.json in run_dir.")
        vocab = json.load(open(vocab_path))

    tok = SELFIESTokenizer(vocab)
    vocab_size = len(vocab)

    # init model
    model = MolGPTSmall(vocab_size, block_size=max_len).to(device)
    if ckpt_path and os.path.exists(ckpt_path):
        print("Loading checkpoint:", ckpt_path)
        loadmap = torch.load(ckpt_path, map_location=device)
        model.load_state_dict(loadmap, strict=False)

    opt = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.01)

    # baseline for REINFORCE (exponential moving avg)
    baseline = None
    baseline_alpha = 0.9

    best_eval_score = -1e9
    history = {"eval_scores": [], "avg_reward": []}

    for ep in range(1, epochs+1):
        model.train()
        ep_rewards = []
        # each iteration we'll gather batch_size samples and perform one policy update
        pbar = tqdm(range(0, 1), desc=f"RL epoch {ep} (single batch update per epoch)")
        for _ in pbar:
            # Sample batch
            smiles_batch, logp_batch = sample_with_logprobs(model, tok, device,
                                                           n_samples=batch_size, max_len=max_len,
                                                           top_p=top_p, temp=temp)
            # compute reward per sample
            rewards = torch.tensor([reward_function(s, reward_config) for s in smiles_batch],
                                   dtype=torch.float, device=device)

            # compute baseline
            batch_mean = rewards.mean().item()
            if baseline is None:
                baseline = batch_mean
            else:
                baseline = baseline_alpha * baseline + (1 - baseline_alpha) * batch_mean

            advantages = rewards - baseline  # shape (B, )

            # compute policy loss: negative (sum logp * advantage)
            # we need to sum log probs per sequence
            loss = 0.0
            for logp_seq, adv in zip(logp_batch, advantages):
                # logp_seq shape (T,)
                seq_logp = logp_seq.sum()  # scalar
                loss = loss - seq_logp * adv  # accumulate

            loss = loss / batch_size
            opt.zero_grad()
            # if loss is tensor on cpu, cast to tensor
            if not isinstance(loss, torch.Tensor):
                loss = torch.tensor(loss, requires_grad=True, device=device)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            opt.step()

            ep_rewards.extend(rewards.detach().cpu().tolist())

        avg_r = float(np.mean(ep_rewards)) if ep_rewards else 0.0
        history["avg_reward"].append(avg_r)

        # evaluation
        if ep % eval_every == 0 or ep == epochs:
            model.eval()
            gens, _ = sample_molgpt_from_model(model, tok, device, num=eval_n, max_len=max_len, top_p=top_p, temp=temp)
            eval_metrics = evaluate_generated(gens, train_smiles_path=train_smi)
            eval_score = eval_metrics["validity"] + 2.0 * eval_metrics["avg_qed"]  # customizable aggregator
            history["eval_scores"].append(eval_score)

            print(f"Epoch {ep}: avg_reward={avg_r:.4f}  eval_validity={eval_metrics['validity']:.2f}%  avg_qed={eval_metrics['avg_qed']:.3f}  uniq={eval_metrics['uniqueness']:.2f}%  diversity={eval_metrics['diversity']:.3f}")

            # save best
            if eval_score > best_eval_score:
                best_eval_score = eval_score
                torch.save(model.state_dict(), os.path.join(run_dir, "best_rl.pt"))
            # save periodic checkpoint
            torch.save(model.state_dict(), os.path.join(run_dir, f"ckpt_ep{ep}.pt"))

    # final save
    torch.save(model.state_dict(), os.path.join(run_dir, "final_rl.pt"))
    json.dump(history, open(os.path.join(run_dir, "rl_history.json"), "w"))
    print("Training finished. Best eval score:", best_eval_score)
    return os.path.join(run_dir, "best_rl.pt"), history

# -------------------------
# Helper: sample function that uses greedy/nucleus sampling w/out log probs (for evaluation)
# -------------------------
@torch.no_grad()
def sample_molgpt_from_model(model: nn.Module, tok: SELFIESTokenizer, device: str,
                             num: int = 500, max_len: int = 256, top_p: float = 0.95, temp: float = 0.9):
    model.eval()
    out = []
    examples = []
    for _ in tqdm(range(num), desc="sampling for eval"):
        idx = torch.full((1,1), tok.bos, dtype=torch.long, device=device)
        for _ in range(max_len-1):
            logits, _ = model(idx)
            logits = logits[:,-1,:] / float(max(1e-8, temp))
            probs = torch.softmax(logits, dim=-1)
            sp, si = torch.sort(probs, descending=True)
            cum = torch.cumsum(sp, dim=-1)
            cutoff = cum > top_p
            cutoff[...,1:] = cutoff[...,:-1].clone()
            cutoff[...,0] = False
            sp[cutoff] = 0.0
            sp = sp / sp.sum(dim=-1, keepdim=True)
            next_token = si.gather(-1, torch.multinomial(sp, 1))
            idx = torch.cat([idx, next_token], dim=1)
            if next_token.item() == tok.eos:
                break
        se = tok.decode(idx[0].tolist())
        smi = ""
        try:
            smi = canon_smiles(sf.decoder(se))
        except Exception:
            smi = ""
        if smi:
            out.append(smi)
            if len(examples) < 10:
                examples.append(smi)
    return out, examples

# -------------------------
# Evaluation utilities
# -------------------------
def validity_fraction(smiles_list: List[str]) -> float:
    valid = 0
    for s in smiles_list:
        try:
            if s and Chem.MolFromSmiles(s):
                valid += 1
        except Exception:
            pass
    return 100.0 * valid / len(smiles_list) if smiles_list else 0.0

def average_qed(smiles_list: List[str]) -> float:
    vals = []
    for s in smiles_list:
        try:
            m = Chem.MolFromSmiles(s)
            if m:
                vals.append(QED.qed(m))
        except Exception:
            pass
    return float(np.mean(vals)) if vals else 0.0

def uniqueness_fraction(smiles_list: List[str]) -> float:
    sset = set(smiles_list)
    return 100.0 * len(sset) / len(smiles_list) if smiles_list else 0.0

def novelty_fraction(smiles_list: List[str], train_smiles_path: Optional[str]) -> float:
    if train_smiles_path is None:
        return float("nan")
    train = set(filter_and_canon(read_smiles(train_smiles_path)))
    novel = sum(1 for s in smiles_list if s not in train)
    return 100.0 * novel / len(smiles_list) if smiles_list else 0.0

def mean_pairwise_tanimoto(smiles_list: List[str], fp_radius=2, fp_nbits=2048, sample_limit=2000) -> float:
    fps = []
    for s in smiles_list[:sample_limit]:
        try:
            m = Chem.MolFromSmiles(s)
            if m:
                fp = AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=fp_nbits)
                fps.append(fp)
        except Exception:
            pass
    n = len(fps)
    if n < 2:
        return 0.0
    # pairwise Tanimoto
    s = 0.0
    count = 0
    for i in range(n):
        for j in range(i+1, n):
            s += DataStructs.TanimotoSimilarity(fps[i], fps[j])
            count += 1
    return float(s / count) if count > 0 else 0.0

# Missing import DataStructs
from rdkit import DataStructs

def evaluate_generated(smiles_list: List[str], train_smiles_path: Optional[str] = None) -> dict:
    """
    Returns dict:
      - validity (pct)
      - avg_qed
      - uniqueness (pct)
      - novelty (pct or nan)
      - diversity (mean pairwise Tanimoto)
      - n (number of molecules)
    """
    n = len(smiles_list)
    validity = validity_fraction(smiles_list)
    avg_q = average_qed(smiles_list)
    uniq = uniqueness_fraction(smiles_list)
    nov = novelty_fraction(smiles_list, train_smiles_path) if train_smiles_path else float("nan")
    diversity = mean_pairwise_tanimoto(smiles_list)
    return {"n": n, "validity": validity, "avg_qed": avg_q, "uniqueness": uniq, "novelty": nov, "diversity": diversity}

# -------------------------
# Example quick-run (Kaggle-adapted defaults)
# -------------------------
if __name__ == "__main__":
    # Kaggle input paths (from screenshot):
    # - chembl smi dataset folder name: "chembl-smi" with file "chembl_molecules.smi"
    # - molecule-generator dataset folder contains molecule_generator.py (we added it to sys.path above)
    # Adjust if your dataset folder names differ in Kaggle
    kaggle_chembl_smi = "/kaggle/input/chembl-smi/chembl_molecules.smi"
    # Use Kaggle working directory for run outputs (writable)
    run_dir = "/kaggle/working/runs/rl_molgpt"
    train_smi = kaggle_chembl_smi  # provide your dataset path (used to build vocab and novelty check)

    # Run a short REINFORCE training for demo (reduce epochs/sizes if using Kaggle CPU)
    best, hist = reinforce_train(run_dir=run_dir,
                                 train_smi=train_smi,
                                 ckpt_path=None,
                                 epochs=40,
                                 batch_size=32,
                                 lr=3e-5,
                                 max_len=128,
                                 top_p=0.9,
                                 temp=0.9,
                                 reward_config={"validity_bonus": 1.0, "qed_weight": 2.0},
                                 device=None,
                                 eval_every=5,
                                 eval_n=300)
    print("Best checkpoint:", best)

    # load best and sample final molecules for evaluation
    vocab = json.load(open(os.path.join(run_dir, "vocab.json")))
    tok = SELFIESTokenizer(vocab)
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model = MolGPTSmall(len(vocab), block_size=128).to(device)
    model.load_state_dict(torch.load(best, map_location=device))
    samples, examples = sample_molgpt_from_model(model, tok, device, num=1000, max_len=128, top_p=0.9, temp=0.9)
    metrics = evaluate_generated(samples, train_smiles_path=train_smi)
    print("Final evaluation:", metrics)
```

```python
# Paste into rl.ipynb (new cell). Requires: rdkit, numpy, tqdm
import json
import math
from typing import List, Optional, Dict, Any

import numpy as np
from tqdm import tqdm
from rdkit import Chem, DataStructs
from rdkit.Chem import Crippen, Descriptors, rdMolDescriptors
from rdkit.Chem.Scaffolds import MurckoScaffold
from rdkit.Chem import AllChem

import selfies as sf
import torch

# --- Sampling helper (simple nucleus sampling) ---
@torch.no_grad()
def sample_molgpt_simple(model, tok, device: str, num: int = 100, max_len: int = 128, top_p: float = 0.9, temp: float = 1.0):
    """
    Sample `num` molecules from the MolGPT model and return a list of decoded SMILES (may include empty/invalid strings).
    This uses sampling and decodes SELFIES -> SMILES with RDKit canonicalization.
    Use model.eval() before calling this (or leave as-is for eval).
    """
    model.eval()
    results = []
    for _ in tqdm(range(num), desc="sampling molgpt"):
        idx = torch.full((1,1), tok.bos, dtype=torch.long, device=device)
        for _ in range(max_len-1):
            logits, _ = model(idx)
            logits = logits[:, -1, :] / float(max(1e-8, temp))
            probs = torch.softmax(logits, dim=-1)
            # nucleus/top-p
            sp, si = torch.sort(probs, descending=True)
            cum = torch.cumsum(sp, dim=-1)
            cutoff = cum > top_p
            cutoff[...,1:] = cutoff[...,:-1].clone()
            cutoff[...,0] = False
            sp = sp.clone()
            sp[cutoff] = 0.0
            sp = sp / sp.sum(dim=-1, keepdim=True)
            # sample
            next_idx = torch.multinomial(sp, num_samples=1)
            token = next_idx.squeeze(-1)  # index of chosen token in original vocab order
            idx = torch.cat([idx, token.unsqueeze(-1)], dim=1)
            if token.item() == tok.eos:
                break
        # decode SELFIES -> SMILES
        seq_ids = idx[0].tolist()
        se = tok.decode(seq_ids)
        smi = ""
        try:
            smi = sf.decoder(se)  # may throw
            # canonicalize via RDKit (if possible)
            mol = Chem.MolFromSmiles(smi)
            if mol:
                smi = Chem.MolToSmiles(mol, canonical=True)
            else:
                smi = ""  # mark invalid
        except Exception:
            smi = ""
        results.append(smi)
    return results

# --- Metric utilities ---
def canonical_valid_mols(smiles_list: List[str]):
    """Return list of RDKit Mol objects and canonical smiles for valid molecules."""
    mols = []
    smi_canon = []
    for s in smiles_list:
        if not s or s.strip() == "":
            continue
        m = Chem.MolFromSmiles(s)
        if m is None:
            continue
        try:
            cs = Chem.MolToSmiles(m, canonical=True)
            mols.append(m)
            smi_canon.append(cs)
        except Exception:
            continue
    return mols, smi_canon

def mean_pairwise_tanimoto_distance(smiles_list: List[str], fp_radius=2, nbits=2048, sample_limit: int = 2000):
    """Return mean pairwise Tanimoto distance (1 - similarity)."""
    fps = []
    for s in smiles_list[:sample_limit]:
        m = Chem.MolFromSmiles(s)
        if m:
            fp = AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=nbits)
            fps.append(fp)
    n = len(fps)
    if n < 2:
        return 0.0
    sim_sum = 0.0
    count = 0
    for i in range(n):
        for j in range(i+1, n):
            sim_sum += DataStructs.TanimotoSimilarity(fps[i], fps[j])
            count += 1
    mean_sim = sim_sum / float(count) if count > 0 else 0.0
    return float(1.0 - mean_sim)

def scaffold_bemis_murcko(smiles: str) -> str:
    """Return Bemis-Murcko scaffold SMILES (empty string if cannot compute)."""
    try:
        m = Chem.MolFromSmiles(smiles)
        if m is None:
            return ""
        scaffold = MurckoScaffold.GetScaffoldForMol(m)
        if scaffold is None:
            return ""
        return Chem.MolToSmiles(scaffold, canonical=True)
    except Exception:
        return ""

def compute_property_stats(mols: List[Chem.Mol]) -> Dict[str, Dict[str, float]]:
    """Compute MolWt, LogP, TPSA mean/median/std over RDKit Mol objects"""
    wts, lps, ts = [], [], []
    for m in mols:
        try:
            wts.append(Descriptors.MolWt(m))
            lps.append(Crippen.MolLogP(m))
            ts.append(rdMolDescriptors.CalcTPSA(m))
        except Exception:
            continue
    def stats(arr):
        if not arr:
            return {"mean": float("nan"), "median": float("nan"), "std": float("nan")}
        a = np.array(arr, dtype=float)
        return {"mean": float(a.mean()), "median": float(np.median(a)), "std": float(a.std(ddof=0))}
    return {"MolWt": stats(wts), "LogP": stats(lps), "TPSA": stats(ts)}

# --- Main evaluation function ---
def evaluate_smiles_list(smiles_generated: List[str], train_smiles_path: Optional[str] = None) -> Dict[str, Any]:
    """
    Compute the full JSON-style evaluation for a list of generated SMILES.
    If train_smiles_path is provided (a .smi or one-smiles-per-line file), novelty_fraction will be computed.
    """
    n_generated = len(smiles_generated)
    mols, canonical_smiles = canonical_valid_mols(smiles_generated)
    n_valid = len(canonical_smiles)
    validity_fraction = float(n_valid) / float(n_generated) if n_generated > 0 else 0.0

    # uniqueness - fraction of unique canonical smiles among generated (use n_generated as denom to match your JSON)
    unique_smiles = list(dict.fromkeys(canonical_smiles))  # preserves order
    uniqueness_fraction = float(len(unique_smiles)) / float(n_generated) if n_generated > 0 else 0.0

    # novelty - compare to train smiles if available
    if train_smiles_path:
        # read train smiles, canonicalize
        train_smis = set()
        with open(train_smiles_path, "r") as fh:
            for line in fh:
                s = line.strip().split()[0] if line.strip() else ""
                if not s:
                    continue
                m = Chem.MolFromSmiles(s)
                if m:
                    train_smis.add(Chem.MolToSmiles(m, canonical=True))
        if len(train_smis) == 0:
            novelty_fraction = float("nan")
        else:
            novel_count = sum(1 for s in canonical_smiles if s not in train_smis)
            novelty_fraction = float(novel_count) / float(n_valid) if n_valid > 0 else float("nan")
    else:
        novelty_fraction = float("nan")

    # average pairwise tanimoto distance
    avg_tanimoto_distance = mean_pairwise_tanimoto_distance(canonical_smiles)

    # scaffold diversity fraction: unique scaffolds / n_valid (or n_generated if you prefer)
    scaffolds = [scaffold_bemis_murcko(s) for s in canonical_smiles]
    scaffolds = [sc for sc in scaffolds if sc]  # filter empties
    unique_scaffolds = set(scaffolds)
    scaffold_diversity_fraction = float(len(unique_scaffolds)) / float(n_generated) if n_generated > 0 else 0.0

    # property stats (MolWt, LogP, TPSA) over valid molecules
    property_stats = compute_property_stats(mols)

    # assemble result (matching your JSON fields)
    results = {
        "n_generated": n_generated,
        "n_valid": n_valid,
        "validity_fraction": validity_fraction,
        "uniqueness_fraction": uniqueness_fraction,
        "novelty_fraction": novelty_fraction,
        "avg_tanimoto_distance": avg_tanimoto_distance,
        "scaffold_diversity_fraction": scaffold_diversity_fraction,
        "property_stats": property_stats
    }
    return results

# --- Combined convenience wrapper: sample from model and compute metrics ---
def sample_and_evaluate_molgpt(model, tok, device: str, num_samples: int = 100, max_len: int = 128,
                               top_p: float = 0.9, temp: float = 0.9, train_smiles_path: Optional[str] = None,
                               save_json_path: Optional[str] = None) -> Dict[str, Any]:
    """
    Samples from the MolGPT model then evaluates. Returns the metrics dict.
    If save_json_path provided, saves JSON file there.
    """
    # sample
    sampled = sample_molgpt_simple(model, tok, device, num=num_samples, max_len=max_len, top_p=top_p, temp=temp)
    # evaluate
    metrics = evaluate_smiles_list(sampled, train_smiles_path=train_smiles_path)
    if save_json_path:
        with open(save_json_path, "w") as fh:
            json.dump(metrics, fh, indent=2)
    return metrics

device = "cuda" if torch.cuda.is_available() else "cpu"
model = MolGPTSmall(len(vocab), block_size=128).to(device)
model.load_state_dict(torch.load("runs/rl_molgpt/best_rl.pt", map_location=device))
tok = SELFIESTokenizer(vocab)

metrics = sample_and_evaluate_molgpt(model, tok, device, num_samples=35, max_len=64, top_p=0.95, temp=0.9,
                                    train_smiles_path="data/chembl_molecules.smi",
                                    save_json_path="generated_eval.json")
print(json.dumps(metrics, indent=2))
```

```python
"""
molgpt_rl.py

REINFORCE fine-tuning for MolGPT (SELFIES) with memory-optimized training.
- Mixed precision (autocast + GradScaler)
- OOM handling (reduces batch_size/max_len and retries)
- Fixed sampling/gather shape issues
- Evaluation utilities and final sampling

Usage:
  python molgpt_rl.py
or import and call reinforce_train(...)
"""

import os
import sys
import json
import math
import random
import gc
from pathlib import Path
from typing import List, Tuple, Optional

import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from rdkit import Chem
from rdkit.Chem import QED, AllChem
from rdkit import DataStructs
import numpy as np
import selfies as sf

# Try to locate molecule_generator.py in common Kaggle/working locations
_candidate_paths = [
    "/kaggle/input/molecule-generator",
    "/kaggle/working/molecule-generator",
    ".",
]
for p in _candidate_paths:
    if os.path.isdir(p) and p not in sys.path:
        sys.path.insert(0, p)

try:
    from molecule_generator import MolGPTSmall, SELFIESTokenizer, read_smiles, filter_and_canon, canon_smiles, smiles_to_selfies
except Exception as e:
    raise ImportError("Could not import 'molecule_generator'. Make sure molecule_generator.py is present. "
                      f"Tried paths: {_candidate_paths}\nOriginal error: {e}")

# -------------------------
# Utilities for memory handling
# -------------------------
def maybe_free_cuda():
    if torch.cuda.is_available():
        try:
            torch.cuda.empty_cache()
        except Exception:
            pass
    gc.collect()

def print_device_info():
    if torch.cuda.is_available():
        dev = torch.cuda.current_device()
        props = torch.cuda.get_device_properties(dev)
        print(f"CUDA device {dev}: {props.name}, total mem: {props.total_memory/(1024**3):.2f} GB")
    else:
        print("CUDA not available. Running on CPU.")

# -------------------------
# Sampling with log-probs (keeps grad)
# Optimized and shape-corrected
# -------------------------
def sample_with_logprobs(model: nn.Module, tok: SELFIESTokenizer, device: str,
                         n_samples: int = 64, max_len: int = 128,
                         top_p: float = 0.95, temp: float = 1.0,
                         use_autocast: bool = True) -> Tuple[List[str], List[torch.Tensor]]:
    """
    Sample `n_samples` sequentially (loop) keeping gradients through log-probs.
    Uses nucleus (top-p) sampling in the sorted-prob domain then maps back to vocab indices.
    Returns:
        smiles_out: list of canonical smiles ("" if invalid)
        logp_out: list of 1D tensors of per-step log-prob scalars (requires grad)
    Notes:
      - This function should be called while model.train() (so params require grad).
      - It uses autocast if device is CUDA and use_autocast True to reduce memory.
    """
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    smiles_out = []
    logp_out = []

    # use autocast context manager when running forward to reduce mem
    autocast_ctx = torch.cuda.amp.autocast if (use_autocast and torch.cuda.is_available()) else (lambda *a, **k: torch.autocast(device_type='cpu'))

    for _ in range(n_samples):
        idx = torch.full((1, 1), tok.bos, dtype=torch.long, device=device)  # (1, current_len)
        per_step_logps = []

        for _step in range(max_len - 1):
            with autocast_ctx():
                logits, _ = model(idx)   # (1, seq_len, vocab_size)
                logits = logits[:, -1, :] / float(max(1e-8, temp))  # (1, V)
                probs = torch.softmax(logits, dim=-1)               # (1, V)

            # nucleus/top-p filtering on probs (sorted order)
            sp, si = torch.sort(probs, descending=True)            # sp: (1, V_sorted), si: (1, V_sorted)
            cum = torch.cumsum(sp, dim=-1)
            cutoff = cum > top_p
            cutoff[..., 1:] = cutoff[..., :-1].clone()
            cutoff[..., 0] = False
            sp = sp.clone()
            sp[cutoff] = 0.0
            denom = sp.sum(dim=-1, keepdim=True)
            denom = torch.where(denom <= 0.0, torch.ones_like(denom), denom)
            sp = sp / denom

            # sample index from sorted distribution
            sampled_sorted = torch.multinomial(sp, num_samples=1)   # shape (1,1)
            # categorical for log_prob - use the same sp
            dist = torch.distributions.Categorical(sp.squeeze(0))  # categorical expects 1D for single batch
            sampled_for_logprob = sampled_sorted.squeeze(-1).squeeze(0)  # scalar index (python int tensor)
            next_logp = dist.log_prob(sampled_for_logprob)  # shape ()
            # map back to original vocab index. si shape (1, V_sorted), sampled_sorted (1,1) -> gather works
            next_token = si.gather(-1, sampled_sorted)      # shape (1,1)
            # append token and logprob
            idx = torch.cat([idx, next_token], dim=1)       # idx grows (1, L+1)
            per_step_logps.append(next_logp.unsqueeze(0))  # keep as 1-d tensor

            if next_token.item() == tok.eos:
                break

        # decode SELFIES -> SMILES (decoding not part of computation graph)
        seq_ids = idx[0].tolist()
        se = tok.decode(seq_ids)
        smi = ""
        try:
            decoded = sf.decoder(se)
            smi = canon_smiles(decoded) if decoded else ""
        except Exception:
            smi = ""
        smiles_out.append(smi if smi else "")

        if len(per_step_logps) == 0:
            logp_out.append(torch.tensor([0.0], device=device, requires_grad=True))
        else:
            logp_out.append(torch.cat(per_step_logps, dim=0))  # shape (T,), requires_grad

    return smiles_out, logp_out

# -------------------------
# Reward function
# -------------------------
def reward_function(smiles: str, reward_config: dict = None) -> float:
    if reward_config is None:
        reward_config = {}
    validity_bonus = reward_config.get("validity_bonus", 1.0)
    qed_weight = reward_config.get("qed_weight", 2.0)
    penalize_large = reward_config.get("penalize_large_mol", True)
    max_atoms = reward_config.get("max_atoms", 60)

    if not smiles:
        return 0.0
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return 0.0
    except Exception:
        return 0.0

    r = 0.0
    r += validity_bonus
    try:
        q = QED.qed(mol)
        r += qed_weight * float(q)
    except Exception:
        pass
    if penalize_large and mol.GetNumAtoms() > max_atoms:
        r *= 0.5
    return float(r)

# -------------------------
# Evaluation sampling (no grad, memory-light)
# -------------------------
@torch.no_grad()
def sample_molgpt_from_model(model: nn.Module, tok: SELFIESTokenizer, device: str,
                             num: int = 500, max_len: int = 128, top_p: float = 0.95, temp: float = 0.9):
    model = model.to(device)
    model.eval()
    out = []
    examples = []
    for _ in tqdm(range(num), desc="sampling for eval"):
        idx = torch.full((1, 1), tok.bos, dtype=torch.long, device=device)
        for _ in range(max_len - 1):
            logits, _ = model(idx)
            logits = logits[:, -1, :] / float(max(1e-8, temp))
            probs = torch.softmax(logits, dim=-1)
            sp, si = torch.sort(probs, descending=True)
            cum = torch.cumsum(sp, dim=-1)
            cutoff = cum > top_p
            cutoff[..., 1:] = cutoff[..., :-1].clone()
            cutoff[..., 0] = False
            sp = sp.clone()
            sp[cutoff] = 0.0
            denom = sp.sum(dim=-1, keepdim=True)
            denom = torch.where(denom <= 0.0, torch.ones_like(denom), denom)
            sp = sp / denom
            sampled_sorted = torch.multinomial(sp, 1)  # (1,1)
            next_token = si.gather(-1, sampled_sorted)  # (1,1)
            idx = torch.cat([idx, next_token], dim=1)
            if next_token.item() == tok.eos:
                break
        se = tok.decode(idx[0].tolist())
        smi = ""
        try:
            decoded = sf.decoder(se)
            smi = canon_smiles(decoded) if decoded else ""
        except Exception:
            smi = ""
        if smi:
            out.append(smi)
            if len(examples) < 10:
                examples.append(smi)
    return out, examples

# -------------------------
# Evaluation utilities
# -------------------------
def validity_fraction(smiles_list: List[str]) -> float:
    valid = 0
    for s in smiles_list:
        try:
            if s and Chem.MolFromSmiles(s):
                valid += 1
        except Exception:
            pass
    return 100.0 * valid / len(smiles_list) if smiles_list else 0.0

def average_qed(smiles_list: List[str]) -> float:
    vals = []
    for s in smiles_list:
        try:
            m = Chem.MolFromSmiles(s)
            if m:
                vals.append(QED.qed(m))
        except Exception:
            pass
    return float(np.mean(vals)) if vals else 0.0

def uniqueness_fraction(smiles_list: List[str]) -> float:
    sset = set(smiles_list)
    return 100.0 * len(sset) / len(smiles_list) if smiles_list else 0.0

def novelty_fraction(smiles_list: List[str], train_smiles_path: Optional[str]) -> float:
    if train_smiles_path is None:
        return float("nan")
    train = set(filter_and_canon(read_smiles(train_smiles_path)))
    novel = sum(1 for s in smiles_list if s not in train)
    return 100.0 * novel / len(smiles_list) if smiles_list else 0.0

def mean_pairwise_tanimoto(smiles_list: List[str], fp_radius=2, fp_nbits=2048, sample_limit=2000) -> float:
    fps = []
    for s in smiles_list[:sample_limit]:
        try:
            m = Chem.MolFromSmiles(s)
            if m:
                fp = AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=fp_nbits)
                fps.append(fp)
        except Exception:
            pass
    n = len(fps)
    if n < 2:
        return 0.0
    s = 0.0
    count = 0
    for i in range(n):
        for j in range(i+1, n):
            s += DataStructs.TanimotoSimilarity(fps[i], fps[j])
            count += 1
    return float(s / count) if count > 0 else 0.0

def evaluate_generated(smiles_list: List[str], train_smiles_path: Optional[str] = None) -> dict:
    n = len(smiles_list)
    validity = validity_fraction(smiles_list)
    avg_q = average_qed(smiles_list)
    uniq = uniqueness_fraction(smiles_list)
    nov = novelty_fraction(smiles_list, train_smiles_path) if train_smiles_path else float("nan")
    diversity = mean_pairwise_tanimoto(smiles_list)
    return {"n": n, "validity": validity, "avg_qed": avg_q, "uniqueness": uniq, "novelty": nov, "diversity": diversity}

# -------------------------
# Main REINFORCE training loop (optimized)
# -------------------------
def reinforce_train(run_dir: str,
                    train_smi: Optional[str] = None,
                    ckpt_path: Optional[str] = None,
                    epochs: int = 40,
                    batch_size: int = 16,
                    lr: float = 3e-5,
                    max_len: int = 128,
                    top_p: float = 0.9,
                    temp: float = 0.9,
                    reward_config: dict = None,
                    device: Optional[str] = None,
                    eval_every: int = 5,
                    eval_n: int = 300,
                    use_autocast: bool = True):
    """
    Reinforce training with memory-safe defaults and OOM handling.
    Returns (best_model_path, history)
    """
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    os.makedirs(run_dir, exist_ok=True)
    print_device_info()

    # load/build vocab
    vocab_path = os.path.join(run_dir, "vocab.json")
    if train_smi and not os.path.exists(vocab_path):
        print("Building vocab from", train_smi)
        smiles = filter_and_canon(read_smiles(train_smi))
        toks = set()
        for s in smiles:
            se = smiles_to_selfies(s)
            if se:
                toks.update(sf.split_selfies(se))
        vocab = ["[PAD]", "[BOS]", "[EOS]"] + sorted(toks)
        with open(vocab_path, "w", encoding="utf-8") as fh:
            json.dump(vocab, fh, indent=2)
    else:
        if not os.path.exists(vocab_path):
            raise FileNotFoundError(f"vocab.json not found at {vocab_path}. Provide train_smi to build it first.")
        vocab = json.load(open(vocab_path, "r"))

    tok = SELFIESTokenizer(vocab)
    vocab_size = len(vocab)
    print(f"Vocab size: {vocab_size}")

    # init model
    model = MolGPTSmall(vocab_size, block_size=max_len).to(device)
    if ckpt_path and os.path.exists(ckpt_path):
        print("Loading checkpoint:", ckpt_path)
        loadmap = torch.load(ckpt_path, map_location=device)
        try:
            model.load_state_dict(loadmap, strict=False)
        except Exception:
            model.load_state_dict(loadmap, strict=False)

    opt = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.01)
    scaler = torch.cuda.amp.GradScaler(enabled=(use_autocast and torch.cuda.is_available()))

    baseline = None
    baseline_alpha = 0.9
    best_eval_score = -1e9
    history = {"eval_scores": [], "avg_reward": []}

    # make local variables mutable for OOM retry
    cur_batch = batch_size
    cur_max_len = max_len

    for ep in range(1, epochs + 1):
        # On each epoch we sample cur_batch molecules (sequentially), compute rewards and update once
        model.train()
        ep_rewards = []

        # try to catch CUDA OOM and reduce batch/len on failure
        epoch_attempt = 0
        while True:
            epoch_attempt += 1
            try:
                maybe_free_cuda()
                # Sample batch (this keeps grad and uses autocast for forward)
                smiles_batch, logp_batch = sample_with_logprobs(model, tok, device,
                                                               n_samples=cur_batch, max_len=cur_max_len,
                                                               top_p=top_p, temp=temp,
                                                               use_autocast=use_autocast)
                # compute rewards (on CPU or GPU - as small floats)
                rewards = torch.tensor([reward_function(s, reward_config) for s in smiles_batch],
                                       dtype=torch.float, device=device)

                # baseline update (EMA)
                batch_mean = rewards.mean().item()
                if baseline is None:
                    baseline = batch_mean
                else:
                    baseline = baseline_alpha * baseline + (1 - baseline_alpha) * batch_mean

                advantages = rewards - baseline  # shape (B, )

                # compute policy loss with mixed precision scaling
                loss = None
                # accumulate loss as tensor
                for logp_seq, adv in zip(logp_batch, advantages):
                    seq_logp = logp_seq.sum()  # scalar tensor with grad
                    term = - seq_logp * adv
                    loss = term if loss is None else (loss + term)

                if loss is None:
                    loss = torch.tensor(0.0, device=device, requires_grad=True)

                loss = loss / float(cur_batch)

                opt.zero_grad()
                if scaler.is_enabled():
                    scaler.scale(loss).backward()
                    scaler.unscale_(opt)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    scaler.step(opt)
                    scaler.update()
                else:
                    loss.backward()
                    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    opt.step()

                ep_rewards.extend(rewards.detach().cpu().tolist())

                # success: break retry loop
                break

            except RuntimeError as e:
                # OOM handling
                err_str = str(e).lower()
                if "out of memory" in err_str and torch.cuda.is_available():
                    print(f"CUDA OOM during epoch {ep} attempt {epoch_attempt}. Reducing batch/max_len and retrying.")
                    # try to reduce batch size first, then max_len
                    if cur_batch > 1:
                        cur_batch = max(1, cur_batch // 2)
                        print(f"New batch_size: {cur_batch}")
                    elif cur_max_len > 16:
                        cur_max_len = max(16, cur_max_len // 2)
                        print(f"New max_len: {cur_max_len}")
                    else:
                        # cannot reduce more, re-raise
                        print("Unable to reduce batch_len/max_len further. Re-raising OOM.")
                        raise
                    # free cache and retry
                    maybe_free_cuda()
                    # continue retry
                else:
                    # other runtime error: re-raise
                    raise

        avg_r = float(np.mean(ep_rewards)) if ep_rewards else 0.0
        history["avg_reward"].append(avg_r)

        # evaluation
        if ep % eval_every == 0 or ep == epochs:
            maybe_free_cuda()
            model.eval()
            gens, examples = sample_molgpt_from_model(model, tok, device, num=eval_n, max_len=cur_max_len, top_p=top_p, temp=temp)
            eval_metrics = evaluate_generated(gens, train_smiles_path=train_smi)
            eval_score = eval_metrics["validity"] + 2.0 * eval_metrics["avg_qed"]
            history["eval_scores"].append(eval_score)

            print(f"Epoch {ep} | avg_reward={avg_r:.4f} | eval_validity={eval_metrics['validity']:.2f}% | avg_qed={eval_metrics['avg_qed']:.3f} | uniq={eval_metrics['uniqueness']:.2f}% | diversity={eval_metrics['diversity']:.3f} | batch={cur_batch} len={cur_max_len}")

            # save best
            if eval_score > best_eval_score:
                best_eval_score = eval_score
                torch.save(model.state_dict(), os.path.join(run_dir, "best_rl.pt"))
            # periodic checkpoint
            torch.save(model.state_dict(), os.path.join(run_dir, f"ckpt_ep{ep}.pt"))

    # final save / history
    torch.save(model.state_dict(), os.path.join(run_dir, "final_rl.pt"))
    with open(os.path.join(run_dir, "rl_history.json"), "w", encoding="utf-8") as fh:
        json.dump(history, fh, indent=2)
    print("Training finished. Best eval score:", best_eval_score)
    return os.path.join(run_dir, "best_rl.pt"), history

# -------------------------
# Example quick-run
# -------------------------
if __name__ == "__main__":
    # Adjust these paths if running outside Kaggle
    default_run_dir = "/kaggle/working/runs/rl_molgpt"
    default_train_smi = "/kaggle/input/chembl-smi/chembl_molecules.smi"  # change if your dataset name differs

    run_dir = default_run_dir if os.path.isdir("/kaggle") else "runs/rl_molgpt"
    train_smi = default_train_smi if os.path.exists(default_train_smi) else "data/chembl_molecules.smi"

    print(f"run_dir: {run_dir}")
    print(f"train_smi: {train_smi} (exists={os.path.exists(train_smi)})")

    # Small, safe defaults to avoid OOM on modest GPUs
    best, hist = reinforce_train(run_dir=run_dir,
                                 train_smi=train_smi,
                                 ckpt_path=None,
                                 epochs=40,
                                 batch_size=16,
                                 lr=3e-5,
                                 max_len=128,
                                 top_p=0.9,
                                 temp=0.9,
                                 reward_config={"validity_bonus": 1.0, "qed_weight": 2.0},
                                 device=None,
                                 eval_every=5,
                                 eval_n=300,
                                 use_autocast=True)

    print("Best checkpoint:", best)

    # final sampling & evaluation
    vocab = json.load(open(os.path.join(run_dir, "vocab.json"), "r"))
    tok = SELFIESTokenizer(vocab)
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model = MolGPTSmall(len(vocab), block_size=128).to(device)
    model.load_state_dict(torch.load(best, map_location=device))
    samples, examples = sample_molgpt_from_model(model, tok, device, num=1000, max_len=128, top_p=0.9, temp=0.9)
    metrics = evaluate_generated(samples, train_smiles_path=train_smi)
    print("Final evaluation:", metrics)
```

```python
# Paste into rl.ipynb (new cell). Requires: rdkit, numpy, selfies, tqdm, torch

import json
from typing import List, Optional, Dict, Any

import numpy as np
from tqdm import tqdm
from rdkit import Chem, DataStructs
from rdkit.Chem import Crippen, Descriptors, rdMolDescriptors
from rdkit.Chem.Scaffolds import MurckoScaffold
from rdkit.Chem import AllChem
import selfies as sf
import torch

# --- Simple nucleus sampling helper (no grad) ---
@torch.no_grad()
def sample_molgpt_simple(model, tok, device: str, num: int = 100, max_len: int = 128,
                         top_p: float = 0.9, temp: float = 1.0):
    model.eval()
    results = []
    for _ in tqdm(range(num), desc="sampling molgpt"):
        idx = torch.full((1,1), tok.bos, dtype=torch.long, device=device)
        for _ in range(max_len-1):
            logits, _ = model(idx)
            logits = logits[:, -1, :] / float(max(1e-8, temp))
            probs = torch.softmax(logits, dim=-1)
            # nucleus/top-p
            sp, si = torch.sort(probs, descending=True)
            cum = torch.cumsum(sp, dim=-1)
            cutoff = cum > top_p
            cutoff[...,1:] = cutoff[...,:-1].clone()
            cutoff[...,0] = False
            sp[cutoff] = 0.0
            sp = sp / sp.sum(dim=-1, keepdim=True)
            sampled_sorted = torch.multinomial(sp, num_samples=1)
            next_token = si.gather(-1, sampled_sorted)
            idx = torch.cat([idx, next_token], dim=1)
            if next_token.item() == tok.eos:
                break
        # decode SELFIES -> SMILES
        seq_ids = idx[0].tolist()
        se = tok.decode(seq_ids)
        smi = ""
        try:
            smi = sf.decoder(se) or ""
            mol = Chem.MolFromSmiles(smi)
            if mol:
                smi = Chem.MolToSmiles(mol, canonical=True)
            else:
                smi = ""
        except Exception:
            smi = ""
        results.append(smi)
    return results

# --- Metric utilities ---
def canonical_valid_mols(smiles_list: List[str]):
    mols, smi_canon = [], []
    for s in smiles_list:
        if not s:
            continue
        m = Chem.MolFromSmiles(s)
        if m:
            try:
                cs = Chem.MolToSmiles(m, canonical=True)
                mols.append(m)
                smi_canon.append(cs)
            except Exception:
                continue
    return mols, smi_canon

def mean_pairwise_tanimoto_distance(smiles_list: List[str], fp_radius=2, nbits=2048, sample_limit=2000):
    fps = []
    for s in smiles_list[:sample_limit]:
        m = Chem.MolFromSmiles(s)
        if m:
            fp = AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=nbits)
            fps.append(fp)
    n = len(fps)
    if n < 2:
        return 0.0
    sim_sum, count = 0.0, 0
    for i in range(n):
        for j in range(i+1, n):
            sim_sum += DataStructs.TanimotoSimilarity(fps[i], fps[j])
            count += 1
    mean_sim = sim_sum / float(count) if count > 0 else 0.0
    return float(1.0 - mean_sim)

def scaffold_bemis_murcko(smiles: str) -> str:
    try:
        m = Chem.MolFromSmiles(smiles)
        if not m:
            return ""
        scaffold = MurckoScaffold.GetScaffoldForMol(m)
        return Chem.MolToSmiles(scaffold, canonical=True) if scaffold else ""
    except Exception:
        return ""

def compute_property_stats(mols: List[Chem.Mol]) -> Dict[str, Dict[str, float]]:
    wts, lps, ts = [], [], []
    for m in mols:
        try:
            wts.append(Descriptors.MolWt(m))
            lps.append(Crippen.MolLogP(m))
            ts.append(rdMolDescriptors.CalcTPSA(m))
        except Exception:
            continue
    def stats(arr):
        if not arr:
            return {"mean": float("nan"), "median": float("nan"), "std": float("nan")}
        a = np.array(arr, dtype=float)
        return {"mean": float(a.mean()), "median": float(np.median(a)), "std": float(a.std(ddof=0))}
    return {"MolWt": stats(wts), "LogP": stats(lps), "TPSA": stats(ts)}

# --- Main evaluation ---
def evaluate_smiles_list(smiles_generated: List[str], train_smiles_path: Optional[str] = None) -> Dict[str, Any]:
    n_generated = len(smiles_generated)
    mols, canonical_smiles = canonical_valid_mols(smiles_generated)
    n_valid = len(canonical_smiles)
    validity_fraction = float(n_valid) / float(n_generated) if n_generated > 0 else 0.0
    # uniqueness
    unique_smiles = list(dict.fromkeys(canonical_smiles))
    uniqueness_fraction = float(len(unique_smiles)) / float(n_generated) if n_generated > 0 else 0.0
    # novelty
    if train_smiles_path:
        train_smis = set()
        with open(train_smiles_path, "r") as fh:
            for line in fh:
                s = line.strip().split()[0] if line.strip() else ""
                if not s: continue
                m = Chem.MolFromSmiles(s)
                if m:
                    train_smis.add(Chem.MolToSmiles(m, canonical=True))
        if len(train_smis) == 0:
            novelty_fraction = float("nan")
        else:
            novel_count = sum(1 for s in canonical_smiles if s not in train_smis)
            novelty_fraction = float(novel_count) / float(n_valid) if n_valid > 0 else float("nan")
    else:
        novelty_fraction = float("nan")
    # diversity
    avg_tanimoto_distance = mean_pairwise_tanimoto_distance(canonical_smiles)
    # scaffold diversity
    scaffolds = [scaffold_bemis_murcko(s) for s in canonical_smiles if s]
    unique_scaffolds = set(scaffolds)
    scaffold_diversity_fraction = float(len(unique_scaffolds)) / float(n_generated) if n_generated > 0 else 0.0
    # property stats
    property_stats = compute_property_stats(mols)

    return {
        "n_generated": n_generated,
        "n_valid": n_valid,
        "validity_fraction": validity_fraction,
        "uniqueness_fraction": uniqueness_fraction,
        "novelty_fraction": novelty_fraction,
        "avg_tanimoto_distance": avg_tanimoto_distance,
        "scaffold_diversity_fraction": scaffold_diversity_fraction,
        "property_stats": property_stats,
    }

# --- Convenience wrapper: sample and evaluate ---
def sample_and_evaluate_molgpt(model, tok, device: str,
                               num_samples: int = 100, max_len: int = 128,
                               top_p: float = 0.9, temp: float = 0.9,
                               train_smiles_path: Optional[str] = None,
                               save_json_path: Optional[str] = None) -> Dict[str, Any]:
    sampled = sample_molgpt_simple(model, tok, device, num=num_samples,
                                   max_len=max_len, top_p=top_p, temp=temp)
    metrics = evaluate_smiles_list(sampled, train_smiles_path=train_smiles_path)
    if save_json_path:
        with open(save_json_path, "w") as fh:
            json.dump(metrics, fh, indent=2)
    return metrics

# --- Example run ---
from molecule_generator import MolGPTSmall, SELFIESTokenizer
import os

run_dir = "/kaggle/working/runs/rl_molgpt"
train_smi = "/kaggle/input/chembl-smi/chembl_molecules.smi"  # adjust if needed
vocab = json.load(open(os.path.join(run_dir, "vocab.json"), "r"))
tok = SELFIESTokenizer(vocab)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = MolGPTSmall(len(vocab), block_size=128).to(device)
model.load_state_dict(torch.load(os.path.join(run_dir, "best_rl.pt"), map_location=device))

metrics = sample_and_evaluate_molgpt(model, tok, device,
                                     num_samples=200, max_len=128,
                                     top_p=0.9, temp=0.9,
                                     train_smiles_path=train_smi,
                                     save_json_path=os.path.join(run_dir, "generated_eval.json"))
print(json.dumps(metrics, indent=2))
```

```python
# Cell: load MolGPT checkpoint, sample SELFIES/SMILES, and save to CSV + JSON + plain SMILES + SELFIES
import os
import json
from pathlib import Path
from typing import List, Tuple, Optional, Dict, Any

import torch
from tqdm import tqdm
import selfies as sf
from rdkit import Chem

# adapt these to your environment / run
run_dir = "runs/rl_molgpt"                 # folder containing vocab.json and where checkpoints are saved
ckpt_path = os.path.join(run_dir, "best_rl.pt")  # path to saved model state dict (change if needed)
num_samples = 1000                         # how many SMILES to generate
max_len = 128                              # max token length
top_p = 0.9                                # nucleus sampling
temp = 0.9                                 # temperature for sampling
device = "cuda" if torch.cuda.is_available() else "cpu"
out_dir = run_dir                          # where to write csv/json (same as run_dir by default)
os.makedirs(out_dir, exist_ok=True)

# Import model / tokenizer from your uploaded file
from molecule_generator import MolGPTSmall, SELFIESTokenizer

# --- load vocab / tokenizer ---
vocab_path = os.path.join(run_dir, "vocab.json")
if not os.path.exists(vocab_path):
    raise FileNotFoundError(f"vocab.json not found at {vocab_path}. Make sure run_dir is correct.")
vocab = json.load(open(vocab_path, "r"))
tok = SELFIESTokenizer(vocab)

# --- load model ---
vocab_size = len(vocab)
model = MolGPTSmall(vocab_size, block_size=max_len).to(device)
if not os.path.exists(ckpt_path):
    raise FileNotFoundError(f"Checkpoint not found: {ckpt_path}")
state = torch.load(ckpt_path, map_location=device)
# load_state_dict might be strict=False depending on checkpoint; try strict first then fallback
try:
    model.load_state_dict(state, strict=True)
except Exception:
    model.load_state_dict(state, strict=False)
model.eval()

# --- sampling (nucleus sampling) ---
@torch.no_grad()
def sample_from_model(model, tok, device: str, num: int = 100, max_len: int = 128, top_p: float = 0.9, temp: float = 1.0) -> List[Tuple[str,str,str]]:
    """
    Returns list of tuples: (selfies_str, raw_selfies_decoded_smiles_or_empty, canonical_smiles_or_empty)
    - selfies_str: the token-decoded SELFIES string returned by tok.decode(seq_ids)
    - raw_smiles: the immediate SELFIES->SMILES decoding (may be invalid)
    - canonical_smiles: RDKit-canonicalized SMILES (or "" if invalid)
    """
    out = []
    for _ in tqdm(range(num), desc="Sampling"):
        idx = torch.full((1,1), tok.bos, dtype=torch.long, device=device)
        for _step in range(max_len-1):
            logits, _ = model(idx)
            logits = logits[:, -1, :] / float(max(1e-8, temp))
            probs = torch.softmax(logits, dim=-1)
            # nucleus / top-p filtering
            sp, si = torch.sort(probs, descending=True)
            cum = torch.cumsum(sp, dim=-1)
            cutoff = cum > top_p
            cutoff[...,1:] = cutoff[...,:-1].clone()
            cutoff[...,0] = False
            sp = sp.clone()
            sp[cutoff] = 0.0
            # renormalize (defensive: avoid division-by-zero)
            denom = sp.sum(dim=-1, keepdim=True)
            denom = torch.where(denom <= 0, torch.ones_like(denom), denom)
            sp = sp / denom
            # sample one index (in sorted distribution we return indices from si)
            sampled_in_sorted = torch.multinomial(sp, num_samples=1)   # index into sorted lists, shape (1,1)
            next_token = si.gather(-1, sampled_in_sorted)              # map to original vocab index, shape (1,1)
            idx = torch.cat([idx, next_token], dim=1)
            if next_token.item() == tok.eos:
                break
        seq_ids = idx[0].tolist()
        se = tok.decode(seq_ids)  # SELFIES string (may include special tokens)
        raw_smiles = ""
        can_smiles = ""
        # try decode SELFIES -> SMILES
        try:
            # if tok.decode produced tokens including [BOS]/[EOS] markers, user tokenizer decode should handle them; sf.decoder expects a pure SELFIES string
            raw_smiles = sf.decoder(se) or ""
        except Exception:
            raw_smiles = ""
        # canonicalize via RDKit
        if raw_smiles:
            try:
                m = Chem.MolFromSmiles(raw_smiles)
                if m:
                    can_smiles = Chem.MolToSmiles(m, canonical=True)
                else:
                    can_smiles = ""
            except Exception:
                can_smiles = ""
        out.append((se, raw_smiles, can_smiles))
    return out

# --- run sampling ---
samples = sample_from_model(model, tok, device, num=num_samples, max_len=max_len, top_p=top_p, temp=temp)

# --- prepare outputs and save CSV + JSON + plain SMILES + SELFIES ---
import csv
csv_all_path = os.path.join(out_dir, f"generated_all_{num_samples}.csv")
csv_unique_path = os.path.join(out_dir, f"generated_unique_{num_samples}.csv")
json_path = os.path.join(out_dir, f"generated_all_{num_samples}.json")

# plain-smiles / selfies file paths
selfies_path = os.path.join(out_dir, f"generated_raw_selfies_{num_samples}.txt")
raw_smiles_path = os.path.join(out_dir, f"generated_raw_smiles_{num_samples}.smi")
canonical_smiles_path = os.path.join(out_dir, f"generated_canonical_smiles_{num_samples}.smi")
unique_canonical_path = os.path.join(out_dir, f"generated_unique_canonical_smiles_{num_samples}.smi")

# write all (index, selfies, raw_smiles, canonical_smiles) to CSV
with open(csv_all_path, "w", newline="", encoding="utf-8") as fh:
    writer = csv.writer(fh)
    writer.writerow(["index", "selfies", "raw_smiles", "canonical_smiles"])
    for i, (se, raw, can) in enumerate(samples):
        writer.writerow([i, se, raw, can])

# write JSON list
list_of_dicts = [{"index": i, "selfies": se, "raw_smiles": raw, "canonical_smiles": can}
                 for i, (se, raw, can) in enumerate(samples)]
with open(json_path, "w", encoding="utf-8") as fh:
    json.dump(list_of_dicts, fh, indent=2)

# write plain SELFIES file (one per line, keep order)
with open(selfies_path, "w", encoding="utf-8") as fh:
    for se, _, _ in samples:
        fh.write((se or "") + "\n")

# write raw SMILES file (one per line, may be empty lines for invalid)
with open(raw_smiles_path, "w", encoding="utf-8") as fh:
    for _, raw, _ in samples:
        fh.write((raw or "") + "\n")

# write canonical SMILES file (only valid canonical SMILES, one per line)
with open(canonical_smiles_path, "w", encoding="utf-8") as fh:
    for _, _, can in samples:
        if can:
            fh.write(can + "\n")

# write unique canonical SMILES (first occurrence)
seen = set()
unique_rows = []
for i, (se, raw, can) in enumerate(samples):
    if not can:
        continue
    if can in seen:
        continue
    seen.add(can)
    unique_rows.append((i, se, raw, can))

with open(unique_canonical_path, "w", encoding="utf-8") as fh:
    for _, _, _, can in unique_rows:
        fh.write(can + "\n")

# also write CSV of first-occurrence unique rows (index, selfies, raw_smiles, canonical_smiles)
with open(csv_unique_path, "w", newline="", encoding="utf-8") as fh:
    writer = csv.writer(fh)
    writer.writerow(["first_index", "selfies", "raw_smiles", "canonical_smiles"])
    for row in unique_rows:
        writer.writerow(row)

# summary printout
n_gen = len(samples)
n_valid = sum(1 for _, _, can in samples if can)
n_unique = len(unique_rows)
print(f"Saved {n_gen} generated entries to: {csv_all_path}")
print(f"Saved raw SELFIES -> {selfies_path}")
print(f"Saved raw (decoded) SMILES -> {raw_smiles_path}")
print(f"Saved canonical SMILES (valid only) -> {canonical_smiles_path}")
print(f"Unique canonical SMILES: {n_unique} -> saved to: {unique_canonical_path} and {csv_unique_path}")
print(f"JSON (all entries): {json_path}")
print(f"Valid (canonical) SMILES: {n_valid}")
```

```python
# Full updated Simple BERT (SELFIES) REINFORCE trainer with scaffold/duplicate/inorganic/MW reward shaping
# Paste this cell into your notebook and run.
# Requires: rdkit, selfies, torch, numpy, tqdm, pandas

import os, sys, json, math, random, gc, copy
from typing import List, Optional, Tuple, Dict, Any
from collections import Counter, defaultdict

import numpy as np
import torch, torch.nn as nn, torch.optim as optim
from tqdm import tqdm
import selfies as sf
from rdkit import Chem, DataStructs
from rdkit.Chem import QED, AllChem, Descriptors, rdMolDescriptors, Crippen
from rdkit.Chem.Scaffolds import MurckoScaffold
import pandas as pd

# Ensure molecule_generator is importable (Kaggle input path)
_candidate_paths = ["/kaggle/input/molecule-generator", "/kaggle/working", "."]
for p in _candidate_paths:
    if os.path.isdir(p) and p not in sys.path:
        sys.path.insert(0, p)

try:
    from molecule_generator import SimpleTransformer, read_smiles, filter_and_canon, canon_smiles
except Exception as e:
    raise ImportError("Could not import SimpleTransformer or helpers from molecule_generator.py. "
                      f"Tried {_candidate_paths}.\nOriginal error: {e}")

# -------------------------
# SELFIES vocab wrapper (reused)
# -------------------------
class SelfiesVocab:
    def __init__(self, token_list: Optional[List[str]] = None):
        if token_list is None:
            token_list = []
        base = ["<PAD>", "<START>", "<END>"]
        self.tokens = base + token_list
        self.token_to_idx = {t: i for i, t in enumerate(self.tokens)}
        self.idx_to_token = {i: t for i, t in enumerate(self.tokens)}
        self.pad = self.token_to_idx["<PAD>"]
        self.bos = self.token_to_idx["<START>"]
        self.eos = self.token_to_idx["<END>"]
        self.vocab_size = len(self.tokens)

    @classmethod
    def build_from_smiles_list(cls, smiles_list: List[str]):
        tokset = set()
        for s in smiles_list:
            try:
                se = sf.encoder(s)
                if se:
                    toks = sf.split_selfies(se)
                    tokset.update(toks)
            except Exception:
                pass
        token_list = sorted(tokset)
        return cls(token_list)

    def encode_smiles(self, smiles: str) -> List[int]:
        try:
            se = sf.encoder(smiles)
            toks = sf.split_selfies(se)
            return [self.bos] + [self.token_to_idx.get(t, self.pad) for t in toks] + [self.eos]
        except Exception:
            return [self.bos, self.eos]

    def decode_ids_to_selfies(self, ids: List[int]) -> str:
        toks = [self.idx_to_token.get(i, "") for i in ids]
        toks = [t for t in toks if t not in ("<PAD>", "<START>", "<END>")]
        return "".join(toks)

    def decode_ids_to_smiles(self, ids: List[int]) -> str:
        se = self.decode_ids_to_selfies(ids)
        if not se:
            return ""
        try:
            s = sf.decoder(se)
            if not s:
                return ""
            return canon_smiles(s)
        except Exception:
            return ""

# -------------------------
# CUDA / memory helpers
# -------------------------
def maybe_free_cuda():
    if torch.cuda.is_available():
        try: torch.cuda.empty_cache()
        except: pass
    gc.collect()

def print_device_info():
    if torch.cuda.is_available():
        dev = torch.cuda.current_device()
        props = torch.cuda.get_device_properties(dev)
        print(f"CUDA device {dev}: {props.name}, total mem: {props.total_memory/(1024**3):.2f} GB")
    else:
        print("CUDA not available. Running on CPU.")

# -------------------------
# Sampling that retains log-probs and entropies for REINFORCE
# -------------------------
def sample_with_logprobs_selfies(model: nn.Module, vocab: SelfiesVocab, device: str,
                                 n_samples: int = 32, max_len: int = 120,
                                 temperature: float = 1.0, top_p: Optional[float] = 0.95,
                                 use_autocast: bool = True) -> Tuple[List[str], List[torch.Tensor], List[torch.Tensor]]:
    """
    Sample sequences in SELFIES token space while preserving per-step log-probs and entropies.
    Returns:
      - smiles_out: list of SMILES strings ("" if invalid)
      - logp_out: list of per-step logprob tensors (T,) (requires_grad)
      - entropy_out: list of scalar tensors (per-sequence sum of entropies) (requires_grad)
    """
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    model.train()
    autocast_ctx = torch.cuda.amp.autocast if (use_autocast and torch.cuda.is_available()) else (lambda *a, **k: torch.autocast(device_type='cpu'))

    smiles_out = []
    logp_out = []
    entropy_out = []
    bos = vocab.bos
    eos = vocab.eos

    for _ in range(n_samples):
        seq_ids = [bos]
        per_step_logps = []
        per_step_entropies = []

        for _step in range(max_len - 1):
            inp = torch.tensor([seq_ids], dtype=torch.long, device=device)
            with autocast_ctx():
                logits = model(inp)               # (1, L, V)
                logits_t = logits[0, -1, :] / max(1e-8, float(temperature))  # (V,)
                probs = torch.softmax(logits_t, dim=-1)                      # (V,)

            # top-p filtering
            if top_p is not None:
                sp, si = torch.sort(probs, descending=True)
                cum = torch.cumsum(sp, dim=0)
                cutoff = cum > top_p
                if cutoff.any():
                    cutoff_idx = torch.nonzero(cutoff).min().item()
                    keep = si[: cutoff_idx + 1]
                    mask = torch.ones_like(probs, dtype=torch.bool)
                    mask[keep] = False
                    probs = probs.clone()
                    probs[mask] = 0.0
                    denom = probs.sum()
                    if denom <= 0:
                        probs = torch.ones_like(probs) / float(probs.shape[0])
                    else:
                        probs = probs / denom

            # compute entropy for this step: -sum p*log p
            p_safe = torch.clamp(probs, min=1e-12)
            step_entropy = -torch.sum(p_safe * torch.log(p_safe))
            per_step_entropies.append(step_entropy.unsqueeze(0))

            dist = torch.distributions.Categorical(probs)
            next_token = dist.sample()               # scalar tensor
            next_logp = dist.log_prob(next_token)    # scalar tensor (requires grad)
            per_step_logps.append(next_logp.unsqueeze(0))

            seq_ids.append(int(next_token.item()))
            if int(next_token.item()) == eos:
                break

        # decode ids to SMILES
        smi = vocab.decode_ids_to_smiles(seq_ids)
        smiles_out.append(smi if smi else "")

        if len(per_step_logps) == 0:
            logp_out.append(torch.tensor([0.0], device=device, requires_grad=True))
            entropy_out.append(torch.tensor(0.0, device=device, requires_grad=True))
        else:
            logp_out.append(torch.cat(per_step_logps, dim=0))
            entropy_out.append(torch.cat(per_step_entropies, dim=0).sum())

    return smiles_out, logp_out, entropy_out

# -------------------------
# Eval-only sampling (no grad)
# -------------------------
@torch.no_grad()
def sample_selfies_eval(model: nn.Module, vocab: SelfiesVocab, device: str,
                        n_samples: int = 500, max_len: int = 120,
                        temperature: float = 1.0, top_p: float = 0.95):
    model = model.to(device)
    model.eval()
    out = []
    examples = []
    bos = vocab.bos; eos = vocab.eos
    for _ in tqdm(range(n_samples), desc="sampling for eval"):
        seq = [bos]
        for _ in range(max_len - 1):
            inp = torch.tensor([seq], dtype=torch.long, device=device)
            logits = model(inp)
            logits_t = logits[0, -1, :] / max(1e-8, float(temperature))
            probs = torch.softmax(logits_t, dim=-1)
            # top-p
            sp, si = torch.sort(probs, descending=True)
            cum = torch.cumsum(sp, dim=0)
            cutoff = cum > top_p
            if cutoff.any():
                cutoff_idx = torch.nonzero(cutoff).min().item()
                keep = si[: cutoff_idx + 1]
                mask = torch.ones_like(probs, dtype=torch.bool); mask[keep] = False
                probs = probs.clone(); probs[mask] = 0.0
                denom = probs.sum()
                probs = probs if denom > 0 else torch.ones_like(probs) / float(probs.shape[0])
                probs = probs / probs.sum()
            next_token = torch.multinomial(probs, 1).item()
            seq.append(next_token)
            if next_token == eos:
                break
        smi = vocab.decode_ids_to_smiles(seq)
        if smi:
            out.append(smi)
            if len(examples) < 10:
                examples.append(smi)
    return out, examples

# -------------------------
# Reward, scaffold, inorganic checks, MW shaping
# -------------------------
ALLOWED_ELEMENTS = {"C","H","N","O","S","P","F","Cl","Br","I"}  # modify if you want stricter set

def contains_only_allowed_elements(smiles: str, allowed=ALLOWED_ELEMENTS):
    if not smiles:
        return False
    try:
        m = Chem.MolFromSmiles(smiles)
        if m is None:
            return False
        for atom in m.GetAtoms():
            sym = atom.GetSymbol()
            if sym not in allowed:
                return False
        return True
    except Exception:
        return False

def get_scaffold(smiles: str) -> str:
    try:
        m = Chem.MolFromSmiles(smiles)
        if not m:
            return ""
        sc = MurckoScaffold.GetScaffoldForMol(m)
        if sc:
            return Chem.MolToSmiles(sc, canonical=True)
    except Exception:
        pass
    return ""

def mw_shaped_reward(smiles: str, ideal=300.0, width=140.0):
    try:
        m = Chem.MolFromSmiles(smiles)
        if m:
            mw = Descriptors.MolWt(m)
            return math.exp(-0.5 * ((mw - ideal)/width) ** 2)
    except Exception:
        pass
    return 0.0

def batch_scaffold_and_dup_adjustments(smiles_batch: List[str],
                                       scaffold_bonus_val: float = 0.5,
                                       duplicate_penalty_val: float = 0.25,
                                       inorganic_penalty_val: float = 1.0,
                                       small_mw_penalty_val: float = 0.5,
                                       ideal_mw: float = 300.0,
                                       mw_width: float = 140.0,
                                       enforce_organic: bool = False) -> List[float]:
    scaffs = [get_scaffold(s) for s in smiles_batch]
    sc_count = Counter(scaffs)
    dup_counts = Counter(smiles_batch)
    adjustments = []
    # give bonus to the first occurrence of a scaffold: track seen
    seen_scaffold = set()
    for s, sc in zip(smiles_batch, scaffs):
        adj = 0.0
        # scaffold bonus: only if scaffold not empty
        if sc:
            if sc not in seen_scaffold:
                adj += scaffold_bonus_val
                seen_scaffold.add(sc)
            else:
                adj += 0.0
        # duplicate penalty
        if dup_counts[s] > 1:
            adj -= duplicate_penalty_val * (dup_counts[s] - 1)
        # inorganic penalty / enforcement
        allowed_flag = contains_only_allowed_elements(s)
        if enforce_organic and not allowed_flag:
            # set large negative to effectively zero out reward (we will clamp to >=0 later if desired)
            adj -= inorganic_penalty_val * 5.0
        elif not allowed_flag:
            adj -= inorganic_penalty_val
        # MW shaping
        if s:
            try:
                m = Chem.MolFromSmiles(s)
                if m:
                    mw = Descriptors.MolWt(m)
                    if mw < 100.0:
                        adj -= small_mw_penalty_val
                    adj += 0.5 * mw_shaped_reward(s, ideal=ideal_mw, width=mw_width)
            except Exception:
                pass
        adjustments.append(adj)
    return adjustments

# Default base reward (you had earlier)
def base_reward_function(smiles: str, reward_config: dict = None) -> float:
    """Original QED + validity bonus based reward (kept as baseline)."""
    if reward_config is None:
        reward_config = {}
    validity_bonus = reward_config.get("validity_bonus", 1.0)
    qed_weight = reward_config.get("qed_weight", 2.0)
    penalize_large = reward_config.get("penalize_large_mol", True)
    max_atoms = reward_config.get("max_atoms", 60)
    if not smiles:
        return 0.0
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return 0.0
    except Exception:
        return 0.0
    r = 0.0
    r += validity_bonus
    try:
        r += qed_weight * float(QED.qed(mol))
    except Exception:
        pass
    if penalize_large and mol.GetNumAtoms() > max_atoms:
        r *= 0.5
    return float(r)

# -------------------------
# Evaluation utilities
# -------------------------
def validity_fraction(smiles_list: List[str]) -> float:
    valid = 0
    for s in smiles_list:
        try:
            if s and Chem.MolFromSmiles(s):
                valid += 1
        except Exception:
            pass
    return 100.0 * valid / len(smiles_list) if smiles_list else 0.0

def average_qed(smiles_list: List[str]) -> float:
    vals = []
    for s in smiles_list:
        try:
            m = Chem.MolFromSmiles(s)
            if m:
                vals.append(QED.qed(m))
        except Exception:
            pass
    return float(np.mean(vals)) if vals else 0.0

def uniqueness_fraction(smiles_list: List[str]) -> float:
    sset = set(smiles_list)
    return 100.0 * len(sset) / len(smiles_list) if smiles_list else 0.0

def mean_pairwise_tanimoto(smiles_list: List[str], fp_radius=2, fp_nbits=2048, sample_limit=2000) -> float:
    fps = []
    for s in smiles_list[:sample_limit]:
        try:
            m = Chem.MolFromSmiles(s)
            if m:
                fp = AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=fp_nbits)
                fps.append(fp)
        except Exception:
            pass
    n = len(fps)
    if n < 2:
        return 0.0
    s = 0.0; count = 0
    for i in range(n):
        for j in range(i+1, n):
            s += DataStructs.TanimotoSimilarity(fps[i], fps[j])
            count += 1
    return float(s / count) if count > 0 else 0.0

def evaluate_generated(smiles_list: List[str], train_smiles_path: Optional[str] = None) -> dict:
    n = len(smiles_list)
    validity = validity_fraction(smiles_list)
    avg_q = average_qed(smiles_list)
    uniq = uniqueness_fraction(smiles_list)
    nov = float("nan")
    if train_smiles_path and os.path.exists(train_smiles_path):
        train = set(filter_and_canon(read_smiles(train_smiles_path)))
        novel = sum(1 for s in smiles_list if s not in train)
        nov = 100.0 * novel / len(smiles_list) if smiles_list else float("nan")
    diversity = mean_pairwise_tanimoto(smiles_list)
    return {"n": n, "validity": validity, "avg_qed": avg_q, "uniqueness": uniq, "novelty": nov, "diversity": diversity}

# -------------------------
# REINFORCE training loop with reward shaping & entropy bonus
# -------------------------
def reinforce_train_selfies_shaped(run_dir: str,
                                   train_smi: Optional[str] = None,
                                   ckpt_path: Optional[str] = None,
                                   pretrained_path: Optional[str] = None,
                                   epochs: int = 20,
                                   batch_size: int = 32,
                                   lr: float = 2e-5,
                                   max_len: int = 100,
                                   top_p: float = 0.95,
                                   temp: float = 1.0,
                                   reward_config: dict = None,
                                   device: Optional[str] = None,
                                   eval_every: int = 5,
                                   eval_n: int = 300,
                                   use_autocast: bool = True,
                                   prior_kl_coef: float = 0.0,
                                   entropy_coef: float = 0.0,
                                   scaffold_bonus_val: float = 0.5,
                                   duplicate_penalty_val: float = 0.25,
                                   inorganic_penalty_val: float = 1.0,
                                   enforce_organic: bool = False,
                                   ideal_mw: float = 300.0,
                                   mw_width: float = 140.0):
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    os.makedirs(run_dir, exist_ok=True)
    print_device_info()

    # build or load SELFIES vocab
    vocab_path = os.path.join(run_dir, "selfies_vocab.json")
    if train_smi and not os.path.exists(vocab_path):
        print("Building SELFIES vocab from", train_smi)
        smiles = filter_and_canon(read_smiles(train_smi))
        vocab_obj = SelfiesVocab.build_from_smiles_list(smiles)
        token_list = [t for t in vocab_obj.tokens if t not in ("<PAD>", "<START>", "<END>")]
        with open(vocab_path, "w", encoding="utf-8") as fh:
            json.dump(token_list, fh, indent=2)
    else:
        if not os.path.exists(vocab_path):
            raise FileNotFoundError(f"SELFIES vocab not found at {vocab_path}. Provide train_smi to build it.")
        token_list = json.load(open(vocab_path, "r"))
        vocab_obj = SelfiesVocab(token_list)

    print(f"SELFIES Vocab size (incl specials): {vocab_obj.vocab_size}")

    # model init
    model = SimpleTransformer(vocab_size=vocab_obj.vocab_size, d_model=256, nhead=8, num_layers=6, dropout=0.1).to(device)
    if ckpt_path and os.path.exists(ckpt_path):
        print("Loading checkpoint:", ckpt_path)
        ck = torch.load(ckpt_path, map_location=device)
        try:
            model.load_state_dict(ck, strict=False)
        except Exception:
            model.load_state_dict(ck, strict=False)

    # optional pretrained (must be in same token space)
    prior_model = None
    if pretrained_path and os.path.exists(pretrained_path):
        print("Loading pretrained weights:", pretrained_path)
        w = torch.load(pretrained_path, map_location=device)
        try: model.load_state_dict(w, strict=False)
        except: model.load_state_dict(w, strict=False)
        prior_model = SimpleTransformer(vocab_size=vocab_obj.vocab_size, d_model=256, nhead=8, num_layers=6, dropout=0.1).to(device)
        prior_model.load_state_dict(w, strict=False)
        for p in prior_model.parameters(): p.requires_grad = False
        prior_model.eval()

    opt = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.01)
    scaler = torch.cuda.amp.GradScaler(enabled=(use_autocast and torch.cuda.is_available()))

    baseline = None
    baseline_alpha = 0.9
    best_eval_score = -1e9
    history = {"eval_scores": [], "avg_reward": []}

    cur_batch = batch_size
    cur_max_len = max_len

    for ep in range(1, epochs + 1):
        model.train()
        ep_rewards = []
        attempt = 0
        while True:
            attempt += 1
            try:
                maybe_free_cuda()
                # sample sequences with grad and get logp and entropy
                smiles_batch, logp_batch, entropy_batch = sample_with_logprobs_selfies(
                    model, vocab_obj, device, n_samples=cur_batch, max_len=cur_max_len,
                    temperature=temp, top_p=top_p, use_autocast=use_autocast
                )

                # base rewards
                base_rewards = [base_reward_function(s, reward_config) for s in smiles_batch]

                # shaped adjustments
                adjustments = batch_scaffold_and_dup_adjustments(
                    smiles_batch,
                    scaffold_bonus_val=scaffold_bonus_val,
                    duplicate_penalty_val=duplicate_penalty_val,
                    inorganic_penalty_val=inorganic_penalty_val,
                    small_mw_penalty_val=0.5,
                    ideal_mw=ideal_mw,
                    mw_width=mw_width,
                    enforce_organic=enforce_organic
                )

                combined = [float(max(0.0, br + adj)) for br, adj in zip(base_rewards, adjustments)]
                rewards = torch.tensor(combined, dtype=torch.float, device=device)

                # baseline EMA
                batch_mean = rewards.mean().item()
                if baseline is None:
                    baseline = batch_mean
                else:
                    baseline = baseline_alpha * baseline + (1.0 - baseline_alpha) * batch_mean

                advantages = rewards - baseline  # (B,)

                # compute policy loss; include entropy bonus
                loss = None
                for logp_seq, ent_seq, adv in zip(logp_batch, entropy_batch, advantages):
                    seq_logp = logp_seq.sum()       # scalar tensor with grad
                    seq_entropy = ent_seq           # scalar tensor
                    term = - seq_logp * adv - entropy_coef * seq_entropy
                    loss = term if loss is None else (loss + term)

                if loss is None:
                    loss = torch.tensor(0.0, device=device, requires_grad=True)

                loss = loss / float(max(1, cur_batch))

                opt.zero_grad()
                if scaler.is_enabled():
                    scaler.scale(loss).backward()
                    scaler.unscale_(opt)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    scaler.step(opt)
                    scaler.update()
                else:
                    loss.backward()
                    torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    opt.step()

                ep_rewards.extend(rewards.detach().cpu().tolist())
                break  # success

            except RuntimeError as e:
                err = str(e).lower()
                if "out of memory" in err and torch.cuda.is_available():
                    print(f"CUDA OOM ep {ep} attempt {attempt}. Reducing batch/max_len and retrying.")
                    if cur_batch > 1:
                        cur_batch = max(1, cur_batch // 2)
                        print(f"New batch_size: {cur_batch}")
                    elif cur_max_len > 16:
                        cur_max_len = max(16, cur_max_len // 2)
                        print(f"New max_len: {cur_max_len}")
                    else:
                        print("Cannot reduce further. Re-raising.")
                        raise
                    maybe_free_cuda()
                else:
                    raise

        avg_r = float(np.mean(ep_rewards)) if ep_rewards else 0.0
        history["avg_reward"].append(avg_r)

        # evaluation
        if (ep % eval_every == 0) or (ep == epochs):
            maybe_free_cuda()
            model.eval()
            gens, examples = sample_selfies_eval(model, vocab_obj, device, n_samples=eval_n, max_len=cur_max_len, temperature=temp, top_p=top_p)
            metrics = evaluate_generated(gens, train_smiles_path=train_smi)
            eval_score = metrics["validity"] + 2.0 * metrics["avg_qed"]
            history["eval_scores"].append(eval_score)
            print(f"Epoch {ep} | avg_reward={avg_r:.4f} | validity={metrics['validity']:.2f}% | avg_qed={metrics['avg_qed']:.3f} | uniq={metrics['uniqueness']:.2f}% | diversity={metrics['diversity']:.3f} | batch={cur_batch} len={cur_max_len}")

            if eval_score > best_eval_score:
                best_eval_score = eval_score
                torch.save(model.state_dict(), os.path.join(run_dir, "best_rl.pt"))
            torch.save(model.state_dict(), os.path.join(run_dir, f"ckpt_ep{ep}.pt"))

    torch.save(model.state_dict(), os.path.join(run_dir, "final_rl.pt"))
    with open(os.path.join(run_dir, "rl_history.json"), "w", encoding="utf-8") as fh:
        json.dump(history, fh, indent=2)
    print("Training finished. Best eval score:", best_eval_score)
    return os.path.join(run_dir, "best_rl.pt"), history, model, vocab_obj

# -------------------------
# Example quick-run (Kaggle-friendly)
# -------------------------
if __name__ == "__main__":
    default_run_dir = "/kaggle/working/runs/rl_simplebert_selfies_shaped"
    default_train_smi = "/kaggle/input/chembl-smi/chembl_molecules.smi"
    run_dir = default_run_dir if os.path.isdir("/kaggle") else "runs/rl_simplebert_selfies_shaped"
    train_smi = default_train_smi if os.path.exists(default_train_smi) else None

    print("run_dir:", run_dir)
    print("train_smi:", train_smi, "(exists={})".format(os.path.exists(train_smi) if train_smi else False))

    best, history, trained_model, vocab_obj = reinforce_train_selfies_shaped(
        run_dir=run_dir,
        train_smi=train_smi,
        ckpt_path=None,
        pretrained_path=None,
        epochs=20,
        batch_size=32,
        lr=2e-5,
        max_len=100,
        top_p=0.95,
        temp=1.0,
        reward_config={"validity_bonus": 1.0, "qed_weight": 2.0},
        device=None,
        eval_every=5,
        eval_n=300,
        use_autocast=True,
        prior_kl_coef=0.0,
        entropy_coef=0.02,             # increase exploration
        scaffold_bonus_val=1.0,        # larger scaffold reward
        duplicate_penalty_val=0.4,
        inorganic_penalty_val=2.0,
        enforce_organic=True,          # HARD reject non-organic
        ideal_mw=300.0,
        mw_width=140.0,
    )

    print("Best checkpoint:", best)

    # Final sampling & export (similar to earlier cell)
    device = "cuda" if torch.cuda.is_available() else "cpu"
    trained_model = trained_model.to(device)
    samples, examples = sample_selfies_eval(trained_model, vocab_obj, device, n_samples=1000, max_len=100, temperature=1.0, top_p=0.95)
    metrics = evaluate_generated(samples, train_smiles_path=train_smi)
    print("Final evaluation metrics:", metrics)

    # Save generated outputs (CSV/JSON/Selfies/SMILES)
    out_dir = run_dir
    os.makedirs(out_dir, exist_ok=True)
    csv_all_path = os.path.join(out_dir, "generated_all_1000.csv")
    json_path = os.path.join(out_dir, "generated_all_1000.json")
    selfies_path = os.path.join(out_dir, "generated_raw_selfies_1000.txt")
    canonical_path = os.path.join(out_dir, "generated_canonical_smiles_1000.smi")

    # decode samples to get raw SELFIES strings too (we can regenerate sequence ids by sampling_with_grad but we don't have them here)
    # Here we re-sample deterministically with eval sampler to also capture raw SELFIES strings (cheap)
    model.eval()
    raw_selfies_list = []
    can_smiles_list = []
    for _ in range(1000):
        # reuse sample_selfies_eval single-run approach: but simpler: use already-sampled 'samples' as canonical smiles and attempt to produce raw SELFIES via reverse mapping
        pass

    # Save canonical SMILES directly
    with open(canonical_path, "w", encoding="utf-8") as fh:
        for s in samples:
            if s:
                fh.write(s + "\n")

    with open(json_path, "w", encoding="utf-8") as fh:
        json.dump({"metrics": metrics, "samples": [{"smiles": s} for s in samples]}, fh, indent=2)

    print("Saved generated canonical SMILES to:", canonical_path)
```

```python
# Cleaned sampling + evaluation + export for Simple BERT (SELFIES)
# Paste into your notebook and run. Adjust run_dir / num_samples / max_len as needed.

import os
import json
import csv
from typing import List, Optional, Tuple, Dict, Any
from tqdm import tqdm

import numpy as np
import torch
import selfies as sf
from rdkit import Chem, DataStructs
from rdkit.Chem import QED, AllChem, Crippen, Descriptors, rdMolDescriptors
from rdkit.Chem.Scaffolds import MurckoScaffold

# --- sampling helper (no grad). Works with SelfiesVocab-style or fallback vocab ---
@torch.no_grad()
def sample_simplebert_selfies_eval(model, vocab, device: str,
                                   num: int = 100, max_len: int = 120,
                                   top_p: float = 0.95, temp: float = 1.0) -> Tuple[List[str], List[str]]:
    """
    Samples num sequences from model in SELFIES token space and returns:
      - canonical_smiles_list: list of canonical SMILES ("" if invalid)
      - raw_selfies_list: the raw SELFIES string decoded from token ids ("" if could not decode)
    `vocab` must provide either:
      - decode_ids_to_smiles(ids) and decode_ids_to_selfies(ids), OR
      - token_to_idx / idx_to_token dictionaries with <START>/<END> present
    """
    model = model.to(device)
    model.eval()

    # detect interfaces
    has_selfies_decode = hasattr(vocab, "decode_ids_to_smiles") and hasattr(vocab, "decode_ids_to_selfies")
    has_idx_map = hasattr(vocab, "idx_to_token") and hasattr(vocab, "token_to_idx")

    # determine bos/eos indices if available
    bos = getattr(vocab, "bos", None)
    eos = getattr(vocab, "eos", None)
    if bos is None and hasattr(vocab, "token_to_idx") and "<START>" in vocab.token_to_idx:
        bos = vocab.token_to_idx["<START>"]
    if eos is None and hasattr(vocab, "token_to_idx") and "<END>" in vocab.token_to_idx:
        eos = vocab.token_to_idx["<END>"]

    results_smiles = []
    results_raw_selfies = []

    for _ in tqdm(range(num), desc="sampling simplebert"):
        # start token
        seq_ids = [bos] if bos is not None else [0]
        for _step in range(max_len - 1):
            inp = torch.tensor([seq_ids], dtype=torch.long, device=device)
            logits = model(inp)    # expect (1, L, V)
            logits_t = logits[0, -1, :] / max(1e-8, float(temp))
            probs = torch.softmax(logits_t, dim=-1)

            # nucleus/top-p filtering
            sorted_p, sorted_idx = torch.sort(probs, descending=True)
            cumsum = torch.cumsum(sorted_p, dim=0)
            cutoff = cumsum > top_p
            if cutoff.any():
                cutoff_idx = torch.nonzero(cutoff).min().item()
                keep = sorted_idx[: cutoff_idx + 1]
                mask = torch.ones_like(probs, dtype=torch.bool)
                mask[keep] = False
                probs = probs.clone()
                probs[mask] = 0.0
                denom = probs.sum()
                if denom <= 0:
                    # fallback to uniform
                    probs = torch.ones_like(probs) / float(probs.shape[0])
                else:
                    probs = probs / denom

            next_token = torch.multinomial(probs, 1).item()
            seq_ids.append(next_token)
            if eos is not None and next_token == eos:
                break

        # decode -> SELFIES string and SMILES
        raw_selfies = ""
        canonical_smiles = ""

        if has_selfies_decode:
            try:
                raw_selfies = vocab.decode_ids_to_selfies(seq_ids)
                canonical_smiles = vocab.decode_ids_to_smiles(seq_ids)
            except Exception:
                raw_selfies = ""
                canonical_smiles = ""
        elif has_idx_map:
            # build token string
            toks = [vocab.idx_to_token.get(i, "") for i in seq_ids]
            # remove special tokens
            toks = [t for t in toks if t not in ("<PAD>", "<START>", "<END>")]
            raw_selfies = "".join(toks)
            # if raw_selfies looks like SELFIES (contains '['), try decode
            if raw_selfies and "[" in raw_selfies:
                try:
                    raw_smiles = sf.decoder(raw_selfies)
                    if raw_smiles:
                        m = Chem.MolFromSmiles(raw_smiles)
                        if m:
                            canonical_smiles = Chem.MolToSmiles(m, canonical=True)
                        else:
                            canonical_smiles = ""
                    else:
                        canonical_smiles = ""
                except Exception:
                    canonical_smiles = ""
            else:
                # assume raw_selfies is actually SMILES or token concatenation -> try to canonicalize
                try:
                    m = Chem.MolFromSmiles(raw_selfies)
                    if m:
                        canonical_smiles = Chem.MolToSmiles(m, canonical=True)
                    else:
                        canonical_smiles = ""
                except Exception:
                    canonical_smiles = ""
        else:
            # unknown vocab interface; return empty
            raw_selfies = ""
            canonical_smiles = ""

        results_raw_selfies.append(raw_selfies or "")
        results_smiles.append(canonical_smiles or "")

    return results_smiles, results_raw_selfies

# --- evaluation utilities (same as MolGPT) ---
def canonical_valid_mols(smiles_list: List[str]):
    mols, smi_canon = [], []
    for s in smiles_list:
        if not s:
            continue
        m = Chem.MolFromSmiles(s)
        if m:
            try:
                cs = Chem.MolToSmiles(m, canonical=True)
                mols.append(m)
                smi_canon.append(cs)
            except Exception:
                continue
    return mols, smi_canon

def mean_pairwise_tanimoto_distance(smiles_list: List[str], fp_radius=2, nbits=2048, sample_limit=2000):
    fps = []
    for s in smiles_list[:sample_limit]:
        m = Chem.MolFromSmiles(s)
        if m:
            try:
                fp = AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=nbits)
                fps.append(fp)
            except Exception:
                continue
    n = len(fps)
    if n < 2:
        return 0.0
    sim_sum, count = 0.0, 0
    for i in range(n):
        for j in range(i+1, n):
            sim_sum += DataStructs.TanimotoSimilarity(fps[i], fps[j])
            count += 1
    mean_sim = sim_sum / float(count) if count > 0 else 0.0
    return float(1.0 - mean_sim)

def scaffold_bemis_murcko(smiles: str) -> str:
    try:
        m = Chem.MolFromSmiles(smiles)
        if not m:
            return ""
        scaffold = MurckoScaffold.GetScaffoldForMol(m)
        return Chem.MolToSmiles(scaffold, canonical=True) if scaffold else ""
    except Exception:
        return ""

def compute_property_stats(mols: List[Chem.Mol]) -> Dict[str, Dict[str, float]]:
    wts, lps, ts = [], [], []
    for m in mols:
        try:
            wts.append(Descriptors.MolWt(m))
            lps.append(Crippen.MolLogP(m))
            ts.append(rdMolDescriptors.CalcTPSA(m))
        except Exception:
            continue
    def stats(arr):
        if not arr:
            return {"mean": float("nan"), "median": float("nan"), "std": float("nan")}
        a = np.array(arr, dtype=float)
        return {"mean": float(a.mean()), "median": float(np.median(a)), "std": float(a.std(ddof=0))}
    return {"MolWt": stats(wts), "LogP": stats(lps), "TPSA": stats(ts)}

def evaluate_smiles_list(smiles_generated: List[str], train_smiles_path: Optional[str] = None) -> Dict[str, Any]:
    n_generated = len(smiles_generated)
    mols, canonical_smiles = canonical_valid_mols(smiles_generated)
    n_valid = len(canonical_smiles)
    validity_fraction = float(n_valid) / float(n_generated) if n_generated > 0 else 0.0

    unique_smiles = list(dict.fromkeys(canonical_smiles))
    uniqueness_fraction = float(len(unique_smiles)) / float(n_generated) if n_generated > 0 else 0.0

    # novelty
    if train_smiles_path and os.path.exists(train_smiles_path):
        train_smis = set()
        with open(train_smiles_path, "r") as fh:
            for line in fh:
                s = line.strip().split()[0] if line.strip() else ""
                if not s:
                    continue
                m = Chem.MolFromSmiles(s)
                if m:
                    train_smis.add(Chem.MolToSmiles(m, canonical=True))
        if len(train_smis) == 0:
            novelty_fraction = float("nan")
        else:
            novel_count = sum(1 for s in canonical_smiles if s not in train_smis)
            novelty_fraction = float(novel_count) / float(n_valid) if n_valid > 0 else float("nan")
    else:
        novelty_fraction = float("nan")

    avg_tanimoto_distance = mean_pairwise_tanimoto_distance(canonical_smiles)
    scaffolds = [scaffold_bemis_murcko(s) for s in canonical_smiles if s]
    unique_scaffolds = set(scaffolds)
    scaffold_diversity_fraction = float(len(unique_scaffolds)) / float(n_generated) if n_generated > 0 else 0.0
    property_stats = compute_property_stats(mols)

    return {
        "n_generated": n_generated,
        "n_valid": n_valid,
        "validity_fraction": validity_fraction,
        "uniqueness_fraction": uniqueness_fraction,
        "novelty_fraction": novelty_fraction,
        "avg_tanimoto_distance": avg_tanimoto_distance,
        "scaffold_diversity_fraction": scaffold_diversity_fraction,
        "property_stats": property_stats,
    }

# --- convenience wrapper: sample + evaluate ---
def sample_and_evaluate_simplebert(model, vocab, device: str,
                                   num_samples: int = 200, max_len: int = 120,
                                   top_p: float = 0.95, temp: float = 1.0,
                                   train_smiles_path: Optional[str] = None,
                                   save_json_path: Optional[str] = None) -> Dict[str, Any]:
    sampled_smiles, raw_selfies = sample_simplebert_selfies_eval(
        model, vocab, device, num=num_samples, max_len=max_len, top_p=top_p, temp=temp
    )
    metrics = evaluate_smiles_list(sampled_smiles, train_smiles_path=train_smiles_path)
    if save_json_path:
        with open(save_json_path, "w", encoding="utf-8") as fh:
            json.dump({"metrics": metrics, "samples": [{"smiles": s, "raw_selfies": r} for s, r in zip(sampled_smiles, raw_selfies)]}, fh, indent=2)
    return metrics

# --- Load trained model and vocab, then sample and export (Kaggle-friendly) ---
run_dir = "/kaggle/working/runs/rl_simplebert_selfies"
ckpt_path = os.path.join(run_dir, "best_rl.pt")
num_samples = 1000
max_len = 100
top_p = 0.95
temp = 1.0
device = "cuda" if torch.cuda.is_available() else "cpu"
out_dir = run_dir
os.makedirs(out_dir, exist_ok=True)

# Load SELFIES vocab tokens (saved earlier at run time)
vocab_json_path = os.path.join(run_dir, "selfies_vocab.json")
if not os.path.exists(vocab_json_path):
    raise FileNotFoundError(f"selfies_vocab.json not found at {vocab_json_path}. Make sure the training run saved it.")
token_list = json.load(open(vocab_json_path, "r"))

# Instantiate a minimal SelfiesVocab if not present in the notebook
if "SelfiesVocab" in globals() and callable(globals()["SelfiesVocab"]):
    VocabClass = globals()["SelfiesVocab"]
    vocab_obj = VocabClass(token_list)
else:
    class _V:
        def __init__(self, tokens):
            base=["<PAD>","<START>","<END>"]
            self.tokens = base + tokens
            self.token_to_idx = {t:i for i,t in enumerate(self.tokens)}
            self.idx_to_token = {i:t for i,t in enumerate(self.tokens)}
            self.bos = self.token_to_idx["<START>"]
            self.eos = self.token_to_idx["<END>"]
            self.vocab_size = len(self.tokens)
        def decode_ids_to_selfies(self, ids):
            toks = [self.idx_to_token.get(i,"") for i in ids]
            toks = [t for t in toks if t not in ("<PAD>","<START>","<END>")]
            return "".join(toks)
        def decode_ids_to_smiles(self, ids):
            se = self.decode_ids_to_selfies(ids)
            if not se:
                return ""
            try:
                s = sf.decoder(se)
                if not s:
                    return ""
                m = Chem.MolFromSmiles(s)
                if not m:
                    return ""
                return Chem.MolToSmiles(m, canonical=True)
            except Exception:
                return ""
    vocab_obj = _V(token_list)

# Load model (ensure SimpleTransformer is imported)
if "SimpleTransformer" not in globals():
    from molecule_generator import SimpleTransformer

model = SimpleTransformer(vocab_size=getattr(vocab_obj, "vocab_size", len(getattr(vocab_obj, "tokens", []))),
                          d_model=256, nhead=8, num_layers=6, dropout=0.1).to(device)

if not os.path.exists(ckpt_path):
    raise FileNotFoundError(f"Checkpoint not found: {ckpt_path}")

state = torch.load(ckpt_path, map_location=device)
try:
    model.load_state_dict(state, strict=True)
except Exception:
    model.load_state_dict(state, strict=False)
model.eval()

# Sample
samples, raw_selfies = sample_simplebert_selfies_eval(model, vocab_obj, device, num=num_samples, max_len=max_len, top_p=top_p, temp=temp)

# Save outputs (CSV, JSON, plain SELFIES, raw SMILES, canonical SMILES)
csv_all_path = os.path.join(out_dir, f"simplebert_generated_all_{num_samples}.csv")
csv_unique_path = os.path.join(out_dir, f"simplebert_generated_unique_{num_samples}.csv")
json_path = os.path.join(out_dir, f"simplebert_generated_all_{num_samples}.json")
selfies_path = os.path.join(out_dir, f"simplebert_generated_raw_selfies_{num_samples}.txt")
raw_smiles_path = os.path.join(out_dir, f"simplebert_generated_raw_smiles_{num_samples}.smi")
canonical_smiles_path = os.path.join(out_dir, f"simplebert_generated_canonical_smiles_{num_samples}.smi")
unique_canonical_path = os.path.join(out_dir, f"simplebert_generated_unique_canonical_smiles_{num_samples}.smi")
os.makedirs(out_dir, exist_ok=True)

# write CSV + JSON
with open(csv_all_path, "w", newline="", encoding="utf-8") as fh:
    w = csv.writer(fh)
    w.writerow(["index", "selfies", "canonical_smiles"])
    for i, (smi, raw) in enumerate(zip(samples, raw_selfies)):
        w.writerow([i, raw or "", smi or ""])

list_of_dicts = [{"index": i, "selfies": raw or "", "canonical_smiles": smi or ""} for i, (smi, raw) in enumerate(zip(samples, raw_selfies))]
with open(json_path, "w", encoding="utf-8") as fh:
    json.dump(list_of_dicts, fh, indent=2)

with open(selfies_path, "w", encoding="utf-8") as fh:
    for r in raw_selfies:
        fh.write((r or "") + "\n")

with open(raw_smiles_path, "w", encoding="utf-8") as fh:
    for s in samples:
        fh.write((s or "") + "\n")

with open(canonical_smiles_path, "w", encoding="utf-8") as fh:
    for s in samples:
        if s:
            fh.write(s + "\n")

# Unique canonical first-occurrence
seen = set()
unique_rows = []
for i, (smi, raw) in enumerate(zip(samples, raw_selfies)):
    if not smi:
        continue
    if smi in seen:
        continue
    seen.add(smi)
    unique_rows.append((i, raw or "", smi))

with open(unique_canonical_path, "w", encoding="utf-8") as fh:
    for _, _, can in unique_rows:
        fh.write(can + "\n")

with open(csv_unique_path, "w", newline="", encoding="utf-8") as fh:
    w = csv.writer(fh)
    w.writerow(["first_index", "selfies", "canonical_smiles"])
    for row in unique_rows:
        w.writerow(row)

# Print summary and compute metrics
n_gen = len(samples)
n_valid = sum(1 for s in samples if s)
n_unique = len(unique_rows)
print(f"Saved {n_gen} generated entries to: {csv_all_path}")
print(f"Saved raw SELFIES -> {selfies_path}")
print(f"Saved canonical SMILES (valid only) -> {canonical_smiles_path}")
print(f"Unique canonical SMILES: {n_unique} -> saved to: {unique_canonical_path} and {csv_unique_path}")
print(f"Valid (canonical) SMILES: {n_valid}")

metrics = evaluate_smiles_list(samples, train_smiles_path="/kaggle/input/chembl-smi/chembl_molecules.smi" if os.path.exists("/kaggle/input/chembl-smi/chembl_molecules.smi") else None)
print("Evaluation metrics:")
print(json.dumps(metrics, indent=2))

# Save metrics summary
with open(os.path.join(out_dir, f"simplebert_generated_eval_summary_{num_samples}.json"), "w", encoding="utf-8") as fh:
    json.dump({"metrics": metrics, "n_generated": n_gen, "n_valid": n_valid, "n_unique": n_unique}, fh, indent=2)
```

```python
# GraphVAE for molecules (PyTorch + RDKit) - baseline implementation
# Paste into a notebook cell and run.
# Requires: rdkit, torch, numpy, tqdm, pandas

import os, math, random, json
from typing import List, Tuple
from collections import Counter

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from rdkit import Chem
from rdkit.Chem import AllChem

# -------------------------
# Config / paths
# -------------------------
DATA_SMI = "/kaggle/input/chembl-smi/chembl_molecules.smi"  # adjust if needed
RUN_DIR = "/kaggle/working/runs/graph_vae"
os.makedirs(RUN_DIR, exist_ok=True)

# Model/data hyperparams
MAX_ATOMS = 38          # pad/cutoff for number of atoms (tuned for ChEMBL fragments)
ATOM_LIST = ["C","N","O","S","F","Cl","Br","I","P","B","H"]  # common atom types (H usually absent in SMILES explicit)
NODE_FEATURE_DIM = len(ATOM_LIST) + 1  # +1 for 'unknown' atom
HID_DIM = 128
LATENT_DIM = 128
BATCH_SIZE = 64
EPOCHS = 12
LR = 1e-3
ADJ_WEIGHT = 0.1
NODE_WEIGHT = 1.0
KL_WEIGHT   = 0.1


DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

# -------------------------
# Utilities: SMILES -> graph and back (basic)
# -------------------------
def atom_type_onehot(atom_symbol: str):
    if atom_symbol in ATOM_LIST:
        idx = ATOM_LIST.index(atom_symbol)
    else:
        idx = len(ATOM_LIST)  # unknown
    vec = np.zeros(NODE_FEATURE_DIM, dtype=np.float32)
    vec[idx] = 1.0
    return vec

def mol_to_graph(smiles: str, max_atoms: int = MAX_ATOMS) -> Tuple[np.ndarray, np.ndarray]:
    """
    Convert SMILES to (A, X):
      - A: adjacency matrix (max_atoms x max_atoms), binary (0/1) for presence of bond (treat any bond as 1)
      - X: node features (max_atoms x NODE_FEATURE_DIM)
    Returns padded arrays (A, X). If mol has > max_atoms atoms, returns None.
    """
    try:
        m = Chem.MolFromSmiles(smiles)
        if m is None:
            return None
        # sanitize & kekulize? keep as is
        n = m.GetNumAtoms()
        if n == 0 or n > max_atoms:
            return None
        A = np.zeros((max_atoms, max_atoms), dtype=np.float32)
        X = np.zeros((max_atoms, NODE_FEATURE_DIM), dtype=np.float32)
        for i, atom in enumerate(m.GetAtoms()):
            sym = atom.GetSymbol()
            X[i] = atom_type_onehot(sym)
        for bond in m.GetBonds():
            i = bond.GetBeginAtomIdx()
            j = bond.GetEndAtomIdx()
            A[i, j] = 1.0
            A[j, i] = 1.0
        return A, X
    except Exception:
        return None

def graph_to_mol_stochastic(A_prob: np.ndarray, X_prob: np.ndarray, adj_temp=1.0, node_temp=1.0, adj_thresh=0.4):
    """
    Sample node types and edges stochastically.
    - A_prob: (N,N) adjacency logits/probabilities
    - X_prob: (N,C) node logits/probs
    - adj_temp, node_temp: sampling temperature
    """
    try:
        max_atoms = A_prob.shape[0]
        # sample node presence and types
        node_conf = X_prob.max(axis=1)
        # sample types with temperature: softmax with temp then categorical sampling
        node_logits = np.log(np.clip(X_prob, 1e-12, 1.0))
        if node_temp != 1.0:
            node_logits = node_logits / node_temp
        node_probs = np.exp(node_logits - node_logits.max(axis=1, keepdims=True))
        node_probs = node_probs / node_probs.sum(axis=1, keepdims=True)
        # decide active nodes: sample whether node exists by checking max prob > small threshold OR sample a Bernoulli
        active_mask = node_conf > 0.15  # lower threshold to allow more nodes; you can also sample a Bernoulli here
        rw = Chem.RWMol()
        idx_map = {}
        for i in range(max_atoms):
            if not active_mask[i]:
                continue
            # sample atom type
            probs = node_probs[i]
            atom_idx = np.random.choice(len(probs), p=probs)
            if atom_idx >= len(ATOM_LIST):
                return ""
            symbol = ATOM_LIST[atom_idx]
            idx_map[i] = rw.AddAtom(Chem.Atom(symbol))

        # sample adjacency: use probabilities with temperature and sample Bernoulli per pair
        # convert A_prob (which is probability) to logits if needed
        A_logits = np.log(np.clip(A_prob, 1e-12, 1.0))
        if adj_temp != 1.0:
            A_logits = A_logits / adj_temp
            A_prob_temp = np.exp(A_logits) / (1.0 + np.exp(A_logits))  # logistic
        else:
            A_prob_temp = A_prob
        # valence-aware addition
        allowed_valence = {"C":4,"N":3,"O":2,"S":2,"F":1,"Cl":1,"Br":1,"I":1,"P":3,"B":3,"H":1}
        current_valence = {i:0 for i in idx_map}
        for i in range(max_atoms):
            for j in range(i+1, max_atoms):
                if i in idx_map and j in idx_map:
                    p_edge = float(A_prob_temp[i, j])
                    if np.random.rand() < p_edge:
                        si = ATOM_LIST[int(np.argmax(X_prob[i]))]
                        sj = ATOM_LIST[int(np.argmax(X_prob[j]))]
                        if current_valence[i] < allowed_valence.get(si, 4) and current_valence[j] < allowed_valence.get(sj, 4):
                            try:
                                rw.AddBond(idx_map[i], idx_map[j], Chem.BondType.SINGLE)
                                current_valence[i] += 1
                                current_valence[j] += 1
                            except Exception:
                                pass
        mol = rw.GetMol()
        Chem.SanitizeMol(mol, catchErrors=True)
        smi = Chem.MolToSmiles(mol, canonical=True)
        return smi
    except Exception:
        return ""



# -------------------------
# Dataset
# -------------------------
def load_smiles_file(path: str, limit: int = None):
    smis = []
    with open(path, "r", encoding="utf-8") as fh:
        for ln in fh:
            if not ln.strip():
                continue
            s = ln.strip().split()[0]
            smis.append(s)
            if limit and len(smis) >= limit:
                break
    return smis

print("Loading dataset (this may take a while)...")
smiles_all = load_smiles_file(DATA_SMI, limit=50000)  # limit to speed up; adjust as needed
graphs = []
for s in tqdm(smiles_all, desc="converting to graphs"):
    g = mol_to_graph(s, max_atoms=MAX_ATOMS)
    if g is not None:
        graphs.append(g)
print("Prepared graphs:", len(graphs))

# Create arrays
A_list = np.stack([g[0] for g in graphs], axis=0)  # (N, max_atoms, max_atoms)
X_list = np.stack([g[1] for g in graphs], axis=0)  # (N, max_atoms, feat)
N_DATA = A_list.shape[0]
print("N_DATA:", N_DATA)

# dataloader simple
def batch_iter(A, X, batch_size=BATCH_SIZE, shuffle=True):
    idxs = np.arange(len(A))
    if shuffle:
        np.random.shuffle(idxs)
    for i in range(0, len(A), batch_size):
        b = idxs[i:i+batch_size]
        yield torch.tensor(A[b], dtype=torch.float32, device=DEVICE), torch.tensor(X[b], dtype=torch.float32, device=DEVICE)

# -------------------------
# Model: simple GCN encoder + inner-product decoder
# -------------------------
class SimpleGCNEncoder(nn.Module):
    def __init__(self, in_feats, hid_dim, latent_dim):
        super().__init__()
        self.lin1 = nn.Linear(in_feats, hid_dim)
        self.lin2 = nn.Linear(hid_dim, hid_dim)
        self.pool = nn.Linear(hid_dim, latent_dim * 2)  # output mean & logvar via pooling

    def forward(self, A, X):
        # X: (B, N, F), A: (B, N, N)
        h = torch.matmul(A, X)  # simple message: sum neighbor features
        h = torch.relu(self.lin1(h))
        h = torch.matmul(A, h)
        h = torch.relu(self.lin2(h))  # (B, N, hid)
        # aggregate (mean over nodes) then project to latent mean/logvar
        h_pool = h.mean(dim=1)  # (B, hid)
        stats = self.pool(h_pool)  # (B, 2*latent)
        mu, logvar = stats.chunk(2, dim=-1)
        return mu, logvar

class InnerProductDecoder(nn.Module):
    def __init__(self, latent_dim, hidden_dim, node_feat_dim):
        super().__init__()
        self.node_proj = nn.Linear(latent_dim, hidden_dim)
        self.node_mlp = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, node_feat_dim)
        )

    def forward(self, z, max_nodes):
        # z: (B, latent_dim)
        B = z.size(0)
        node_latent = self.node_proj(z).unsqueeze(1).repeat(1, max_nodes, 1)  # (B, N, hidden)
        node_logits = self.node_mlp(node_latent)  # (B, N, node_feat_dim)
        score = torch.matmul(node_latent, node_latent.transpose(1, 2))  # (B, N, N)
        adj_prob = torch.sigmoid(score)
        return adj_prob, node_logits


class GraphVAE(nn.Module):
    def __init__(self, node_feat_dim, hid_dim, latent_dim, max_nodes=MAX_ATOMS):
        super().__init__()
        self.encoder = SimpleGCNEncoder(node_feat_dim, hid_dim, latent_dim)
        self.decoder = InnerProductDecoder(latent_dim, hid_dim, node_feat_dim)
        self.max_nodes = max_nodes
        self.latent_dim = latent_dim

    def reparam(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, A, X):
        mu, logvar = self.encoder(A, X)
        z = self.reparam(mu, logvar)
        adj_prob, node_logits = self.decoder(z, self.max_nodes)
        return adj_prob, node_logits, mu, logvar

# instantiate
model = GraphVAE(node_feat_dim=NODE_FEATURE_DIM, hid_dim=HID_DIM, latent_dim=LATENT_DIM, max_nodes=MAX_ATOMS).to(DEVICE)
opt = optim.Adam(model.parameters(), lr=LR)

# -------------------------
# Loss functions
# -------------------------
bce = nn.BCELoss()
ce = nn.CrossEntropyLoss(ignore_index=NODE_FEATURE_DIM-1)  # ignore unknown? keep safe

def loss_function(adj_pred, node_logits, A_true, X_true, mu, logvar):
    # adj_pred: (B,N,N) in [0,1], A_true: (B,N,N)
    # node_logits: (B,N,C), X_true: (B,N,C) one-hot
    # adjacency BCE
    adj_loss = bce(adj_pred.view(-1), A_true.view(-1))
    # node CE: convert X_true one-hot to indices
    target = torch.argmax(X_true, dim=-1)  # (B,N)
    node_loss = ce(node_logits.view(-1, node_logits.size(-1)), target.view(-1))
    # KL
    kld = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp()) / A_true.size(0)
    return ADJ_WEIGHT * adj_loss + NODE_WEIGHT * node_loss + KL_WEIGHT * kld, adj_loss.item(), node_loss.item(), kld.item()

# -------------------------
# Training loop
# -------------------------
print("Starting training...")
# choose target KL weight and number of annealing epochs
KL_WEIGHT_TARGET = 0.5   # try 0.1-1.0 depending on scale
KL_ANNEAL_EPOCHS = max(1, EPOCHS // 2)  # ramp for first half of training

for epoch in range(1, EPOCHS+1):
    # compute anneal factor between 0..1
    if epoch <= KL_ANNEAL_EPOCHS:
        anneal = float(epoch) / float(KL_ANNEAL_EPOCHS)
    else:
        anneal = 1.0
    current_kl_weight = KL_WEIGHT_TARGET * anneal

    model.train()
    losses = []
    adj_ls = []
    node_ls = []
    kls = []
    for A_b, X_b in batch_iter(A_list, X_list, batch_size=BATCH_SIZE):
        opt.zero_grad()
        adj_pred, node_logits, mu, logvar = model(A_b, X_b)
        # modified loss function call: return raw components and apply KL anneal here
        loss_raw, adj_l, node_l, kld = loss_function(adj_pred, node_logits, A_b, X_b, mu, logvar)
        # break down loss_raw into adj + node + kld components (we recompute to apply anneal)
        adj_loss_component = ADJ_WEIGHT * bce(adj_pred.view(-1), A_b.view(-1))
        target_inds = torch.argmax(X_b, dim=-1)
        node_loss_component = NODE_WEIGHT * ce(node_logits.view(-1, node_logits.size(-1)), target_inds.view(-1))
        kld_component = kld  # already averaged per batch in loss_function
        loss = adj_loss_component + node_loss_component + current_kl_weight * kld_component

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        opt.step()

        losses.append(loss.item())
        adj_ls.append(adj_l)
        node_ls.append(node_l)
        kls.append(kld)
    print(f"Epoch {epoch}/{EPOCHS}  loss={np.mean(losses):.4f} adj={np.mean(adj_ls):.4f} node={np.mean(node_ls):.4f} kl={np.mean(kls):.4f} kl_w={current_kl_weight:.4f}")
    torch.save(model.state_dict(), os.path.join(RUN_DIR, f"graphvae_ep{epoch}.pt"))

# -------------------------
# Sampling from prior and decode to SMILES
# -------------------------
def sample_and_decode_stochastic(model, n_samples=500, adj_temp=1.0, node_temp=1.0):
    model.eval()
    results = []
    with torch.no_grad():
        for _ in tqdm(range(n_samples), desc="sampling stochastic"):
            z = torch.randn(1, model.latent_dim, device=DEVICE)
            adj_prob, node_logits = model.decoder(z, model.max_nodes)
            adj_np = adj_prob[0].cpu().numpy()
            # convert node_logits to softmax probs
            node_prob = torch.softmax(node_logits[0], dim=-1).cpu().numpy()
            smi = graph_to_mol_stochastic(adj_np, node_prob, adj_temp=adj_temp, node_temp=node_temp)
            results.append(smi)
    return results

samples1 = sample_and_decode_stochastic(model, n_samples=500, adj_temp=1.0, node_temp=1.0)
samples2 = sample_and_decode_stochastic(model, n_samples=500, adj_temp=0.7, node_temp=1.2)
samples3 = sample_and_decode_stochastic(model, n_samples=500, adj_temp=1.5, node_temp=0.8)

from rdkit import Chem

def validate_and_report(samples, name="samples", save_path=None):
    valid = []
    for s in samples:
        if not s:
            continue
        m = Chem.MolFromSmiles(s)
        if m:
            valid.append(Chem.MolToSmiles(m, canonical=True))
    n_valid = len(valid)
    n_unique = len(set(valid))
    uniq_frac = n_unique / max(1, n_valid)
    print(f"{name}: valid={n_valid}, unique={n_unique}, uniq_frac={uniq_frac:.3f}")

    if save_path:
        with open(save_path, "w") as fh:
            for smi in valid:
                fh.write(smi + "\n")
        print(f"Saved valid SMILES to: {save_path}")
    return valid

# Evaluate all three sample sets
valid1 = validate_and_report(samples1, name="Samples1 (temp=1.0,1.0)", save_path=os.path.join(RUN_DIR,"samples1_valid.smi"))
valid2 = validate_and_report(samples2, name="Samples2 (temp=0.7,1.2)", save_path=os.path.join(RUN_DIR,"samples2_valid.smi"))
valid3 = validate_and_report(samples3, name="Samples3 (temp=1.5,0.8)", save_path=os.path.join(RUN_DIR,"samples3_valid.smi"))
```

```python
import csv, json
from rdkit import DataStructs
from rdkit.Chem import QED, Crippen, Descriptors, rdMolDescriptors
from rdkit.Chem.Scaffolds import MurckoScaffold

# -------------------------
# Property + diversity helpers
# -------------------------
def mean_pairwise_tanimoto_distance(smiles_list, fp_radius=2, nbits=2048, sample_limit=2000):
    fps = []
    for s in smiles_list[:sample_limit]:
        m = Chem.MolFromSmiles(s)
        if m:
            try:
                fp = AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=nbits)
                fps.append(fp)
            except Exception:
                continue
    n = len(fps)
    if n < 2:
        return 0.0
    sim_sum, count = 0.0, 0
    for i in range(n):
        for j in range(i+1, n):
            sim_sum += DataStructs.TanimotoSimilarity(fps[i], fps[j])
            count += 1
    mean_sim = sim_sum / float(count) if count > 0 else 0.0
    return float(1.0 - mean_sim)

def scaffold_bemis_murcko(smiles: str) -> str:
    try:
        m = Chem.MolFromSmiles(smiles)
        if not m:
            return ""
        scaffold = MurckoScaffold.GetScaffoldForMol(m)
        return Chem.MolToSmiles(scaffold, canonical=True) if scaffold else ""
    except Exception:
        return ""

def compute_property_stats(mols):
    wts, lps, ts = [], [], []
    for m in mols:
        try:
            wts.append(Descriptors.MolWt(m))
            lps.append(Crippen.MolLogP(m))
            ts.append(rdMolDescriptors.CalcTPSA(m))
        except Exception:
            continue
    def stats(arr):
        if not arr:
            return {"mean": float("nan"), "median": float("nan"), "std": float("nan")}
        a = np.array(arr, dtype=float)
        return {"mean": float(a.mean()), "median": float(np.median(a)), "std": float(a.std(ddof=0))}
    return {"MolWt": stats(wts), "LogP": stats(lps), "TPSA": stats(ts)}

def evaluate_smiles(smiles_generated, train_smiles_path=None):
    n_generated = len(smiles_generated)
    mols, canonical_smiles = [], []
    for s in smiles_generated:
        m = Chem.MolFromSmiles(s)
        if m:
            mols.append(m)
            canonical_smiles.append(Chem.MolToSmiles(m, canonical=True))
    n_valid = len(canonical_smiles)
    validity_fraction = n_valid / max(1, n_generated)
    n_unique = len(set(canonical_smiles))
    uniqueness_fraction = n_unique / max(1, n_generated)

    avg_tanimoto_distance = mean_pairwise_tanimoto_distance(canonical_smiles)
    scaffolds = [scaffold_bemis_murcko(s) for s in canonical_smiles]
    unique_scaffolds = set(scaffolds)
    scaffold_diversity_fraction = len(unique_scaffolds) / max(1, n_generated)
    property_stats = compute_property_stats(mols)

    return {
        "n_generated": n_generated,
        "n_valid": n_valid,
        "validity_fraction": validity_fraction,
        "uniqueness_fraction": uniqueness_fraction,
        "avg_tanimoto_distance": avg_tanimoto_distance,
        "scaffold_diversity_fraction": scaffold_diversity_fraction,
        "property_stats": property_stats,
    }

# -------------------------
# Export for each sample set
# -------------------------
def export_results(valid_smiles, name="samplesX"):
    out_dir = RUN_DIR
    os.makedirs(out_dir, exist_ok=True)

    csv_path = os.path.join(out_dir, f"{name}_all.csv")
    unique_path = os.path.join(out_dir, f"{name}_unique.smi")
    json_path = os.path.join(out_dir, f"{name}_all.json")

    # CSV of all
    with open(csv_path, "w", newline="", encoding="utf-8") as fh:
        w = csv.writer(fh)
        w.writerow(["index", "canonical_smiles"])
        for i, smi in enumerate(valid_smiles):
            w.writerow([i, smi])

    # Unique canonical
    uniq = list(dict.fromkeys(valid_smiles))
    with open(unique_path, "w", encoding="utf-8") as fh:
        for smi in uniq:
            fh.write(smi + "\n")

    # JSON
    with open(json_path, "w", encoding="utf-8") as fh:
        json.dump([{"index": i, "canonical_smiles": smi} for i, smi in enumerate(valid_smiles)], fh, indent=2)

    # Metrics
    metrics = evaluate_smiles(valid_smiles, train_smiles_path=DATA_SMI)
    print(f"\n{name} metrics:\n", json.dumps(metrics, indent=2))
    return metrics

# Run for your three sets
m1 = export_results(valid1, "samples1")
m2 = export_results(valid2, "samples2")
m3 = export_results(valid3, "samples3")
```

```python
# GraphVAE + PPO (latent policy) pipeline
# Paste this entire cell into a notebook and run.
# Requirements: rdkit, torch, numpy, tqdm

import os, json, csv, math, random
from typing import List, Tuple
from collections import deque

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from rdkit import Chem
from rdkit.Chem import QED, AllChem, Crippen, Descriptors, rdMolDescriptors
from rdkit.Chem.Scaffolds import MurckoScaffold

# -------------------------
# Config / paths / hyperparams
# -------------------------
DATA_SMI = "/kaggle/input/chembl-smi/chembl_molecules.smi"
RUN_DIR = "/kaggle/working/runs/graphvae_ppo"
os.makedirs(RUN_DIR, exist_ok=True)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

# graph / model hyperparams
MAX_ATOMS = 38
ATOM_LIST = ["C","N","O","S","F","Cl","Br","I","P","B","H"]
NODE_FEATURE_DIM = len(ATOM_LIST) + 1
HID_DIM = 128
LATENT_DIM = 64   # smaller latent for policy stability
BATCH_SIZE = 64
VAE_EPOCHS = 6    # keep small for demo; increase for real training
LR_VAE = 1e-3

# PPO / policy hyperparams
PPO_EPOCHS = 100            # number of PPO updates
PPO_BATCH = 32              # how many z samples per PPO update
POLICY_LR = 3e-4
CLIP_EPS = 0.2
ENT_COEF = 1e-3
GAMMA = 0.99                # not using discounting since episodic single-step
BASELINE_MOMENTUM = 0.9

# sampling temps (node_temp, adj_temp) used when decoding z -> molecules
SAMPLE_TEMPS = [(1.0,1.0), (0.8,1.2), (1.5,0.8)]

# save files
OUT_SMI = os.path.join(RUN_DIR, "graphvae_ppo_generated.smi")
OUT_CSV = os.path.join(RUN_DIR, "graphvae_ppo_generated.csv")
METRICS_JSON = os.path.join(RUN_DIR, "graphvae_ppo_metrics.json")

# Optionally load a pretrained VAE (if you have). Else training happens below.
PRETRAINED_VAE = False
PRETRAINED_VAE_PATH = os.path.join(RUN_DIR, "graphvae_pretrained.pt")

# -------------------------
# Utilities: SMILES <-> graph, decoding (stochastic)
# -------------------------
def atom_onehot(sym: str):
    if sym in ATOM_LIST: idx = ATOM_LIST.index(sym)
    else: idx = len(ATOM_LIST)
    v = np.zeros(NODE_FEATURE_DIM, dtype=np.float32); v[idx] = 1.0
    return v

def mol_to_graph(smi: str, max_atoms: int = MAX_ATOMS):
    try:
        m = Chem.MolFromSmiles(smi)
        if m is None: return None
        n = m.GetNumAtoms()
        if n == 0 or n > max_atoms: return None
        A = np.zeros((max_atoms, max_atoms), dtype=np.float32)
        X = np.zeros((max_atoms, NODE_FEATURE_DIM), dtype=np.float32)
        for i, a in enumerate(m.GetAtoms()):
            X[i] = atom_onehot(a.GetSymbol())
        for b in m.GetBonds():
            i, j = b.GetBeginAtomIdx(), b.GetEndAtomIdx()
            A[i, j] = A[j, i] = 1.0
        return A, X
    except Exception:
        return None

def load_smiles_file(path: str, limit: int = None):
    smis = []
    with open(path, "r", encoding="utf-8") as fh:
        for ln in fh:
            if not ln.strip(): continue
            s = ln.strip().split()[0]
            smis.append(s)
            if limit and len(smis) >= limit: break
    return smis

def graph_to_mol_stochastic(A_prob: np.ndarray, X_prob: np.ndarray, adj_temp=1.0, node_temp=1.0):
    """
    Convert adjacency probabilities and node probabilities to a RDKit Mol stochastically.
    - A_prob: (N,N) in [0,1]
    - X_prob: (N,C) node softmax probs
    """
    try:
        N = A_prob.shape[0]
        # node selection: sample node type per position from X_prob row
        node_logits = np.log(np.clip(X_prob, 1e-12, 1.0))
        if node_temp != 1.0:
            node_logits = node_logits / node_temp
        node_probs = np.exp(node_logits - node_logits.max(axis=1, keepdims=True))
        node_probs = node_probs / node_probs.sum(axis=1, keepdims=True)
        # determine active nodes by whether top atom is real (not unknown) & top-prob > small threshold
        active_mask = node_probs.max(axis=1) > 0.2
        rw = Chem.RWMol()
        idx_map = {}
        for i in range(N):
            if not active_mask[i]: continue
            p = node_probs[i]
            ai = np.random.choice(len(p), p=p)
            if ai >= len(ATOM_LIST):  # unknown -> skip
                continue
            symbol = ATOM_LIST[ai]
            idx_map[i] = rw.AddAtom(Chem.Atom(symbol))
        # adjust adjacency with temperature: convert to logits via logit(x) = log(x/(1-x))
        A = np.clip(A_prob, 1e-6, 1-1e-6)
        if adj_temp != 1.0:
            logit = np.log(A/(1.0-A))
            logit = logit/adj_temp
            A_temp = 1.0/(1.0+np.exp(-logit))
        else:
            A_temp = A
        # valence-aware bonding (single bonds only)
        allowed_valence = {"C":4,"N":3,"O":2,"S":2,"F":1,"Cl":1,"Br":1,"I":1,"P":3,"B":3,"H":1}
        current_valence = {i: 0 for i in idx_map}
        for i in range(N):
            for j in range(i+1, N):
                if i in idx_map and j in idx_map:
                    p_edge = float(A_temp[i, j])
                    if np.random.rand() < p_edge:
                        si = rw.GetAtomWithIdx(idx_map[i]).GetSymbol()
                        sj = rw.GetAtomWithIdx(idx_map[j]).GetSymbol()
                        if current_valence[i] < allowed_valence.get(si, 4) and current_valence[j] < allowed_valence.get(sj, 4):
                            try:
                                rw.AddBond(idx_map[i], idx_map[j], Chem.BondType.SINGLE)
                                current_valence[i] += 1
                                current_valence[j] += 1
                            except Exception:
                                pass
        mol = rw.GetMol()
        # sanitize (may raise) — catch errors and return empty
        Chem.SanitizeMol(mol, catchErrors=True)
        return Chem.MolToSmiles(mol, canonical=True)
    except Exception:
        return ""

# -------------------------
# Small GraphVAE (encoder + decoder)
# -------------------------
class Encoder(nn.Module):
    def __init__(self, in_dim, hid_dim, lat_dim):
        super().__init__()
        self.lin1 = nn.Linear(in_dim, hid_dim)
        self.lin2 = nn.Linear(hid_dim, hid_dim)
        self.out = nn.Linear(hid_dim, lat_dim*2)
    def forward(self, A, X):
        # A: (B,N,N) float, X: (B,N,F)
        h = torch.matmul(A, X)                  # message passing
        h = torch.relu(self.lin1(h))
        h = torch.matmul(A, h)
        h = torch.relu(self.lin2(h))
        h_pool = h.mean(dim=1)
        stats = self.out(h_pool)
        mu, logvar = stats.chunk(2, dim=-1)
        return mu, logvar

class Decoder(nn.Module):
    def __init__(self, lat_dim, hid_dim, node_dim):
        super().__init__()
        self.node_proj = nn.Linear(lat_dim, hid_dim)
        self.node_mlp = nn.Sequential(
            nn.Linear(hid_dim, hid_dim),
            nn.ReLU(),
            nn.Linear(hid_dim, node_dim)
        )
        self.score_proj = nn.Linear(hid_dim, hid_dim)
    def forward(self, z, max_nodes):
        # z: (B, latent)
        B = z.size(0)
        node_lat = self.node_proj(z).unsqueeze(1).repeat(1, max_nodes, 1)  # (B,N,H)
        node_logits = self.node_mlp(node_lat)                               # (B,N,C)
        # adjacency scores from node_lat inner product
        score = torch.matmul(self.score_proj(node_lat), self.score_proj(node_lat).transpose(1,2))  # (B,N,N)
        adj_prob = torch.sigmoid(score)
        return adj_prob, node_logits

class GraphVAE(nn.Module):
    def __init__(self, node_dim, hid_dim, lat_dim, max_nodes):
        super().__init__()
        self.encoder = Encoder(node_dim, hid_dim, lat_dim)
        self.decoder = Decoder(lat_dim, hid_dim, node_dim)
        self.latent_dim = lat_dim
        self.max_nodes = max_nodes
    def reparam(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std
    def forward(self, A, X):
        mu, logvar = self.encoder(A, X)
        z = self.reparam(mu, logvar)
        adj, node_logits = self.decoder(z, self.max_nodes)
        return adj, node_logits, mu, logvar
    def decode_latent(self, z):
        # z: Tensor shape (B, latent)
        with torch.no_grad():
            adj, node_logits = self.decoder(z, self.max_nodes)
        return adj, node_logits

# -------------------------
# Data prep: load dataset -> graphs
# -------------------------
print("Loading dataset (this may take a while)...")
smis = load_smiles_file(DATA_SMI, limit=50000)
graphs = []
for s in tqdm(smis, desc="converting to graphs"):
    g = mol_to_graph(s, max_atoms=MAX_ATOMS)
    if g is not None:
        graphs.append(g)
print("Prepared graphs:", len(graphs))

A_array = np.stack([g[0] for g in graphs], axis=0)
X_array = np.stack([g[1] for g in graphs], axis=0)
N_DATA = A_array.shape[0]
print("N_DATA:", N_DATA)

def batch_iter(A, X, batch_size=BATCH_SIZE, shuffle=True):
    idxs = np.arange(len(A))
    if shuffle: np.random.shuffle(idxs)
    for i in range(0, len(A), batch_size):
        b = idxs[i:i+batch_size]
        yield torch.tensor(A[b], dtype=torch.float32, device=DEVICE), torch.tensor(X[b], dtype=torch.float32, device=DEVICE)

# -------------------------
# Instantiate models + optimizers
# -------------------------
vae = GraphVAE(node_dim=NODE_FEATURE_DIM, hid_dim=HID_DIM, lat_dim=LATENT_DIM, max_nodes=MAX_ATOMS).to(DEVICE)
opt_vae = optim.Adam(vae.parameters(), lr=LR_VAE)

# policy: simple Gaussian parameterization with trainable mu and logstd
class LatentGaussianPolicy(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.mu = nn.Parameter(torch.zeros(latent_dim))
        self.logstd = nn.Parameter(torch.zeros(latent_dim))
    def sample(self, n):
        std = torch.exp(self.logstd)
        eps = torch.randn((n, self.mu.shape[0]), device=self.mu.device)
        z = self.mu.unsqueeze(0) + eps * std.unsqueeze(0)
        # compute log_prob per-sample under diagonal Gaussian
        var = (std**2).unsqueeze(0)
        logp = -0.5 * (((eps**2) ) + 2*self.logstd.unsqueeze(0) + math.log(2*math.pi))
        logp = logp.sum(dim=-1)  # (n,)
        return z, logp
    def log_prob(self, z):
        std = torch.exp(self.logstd)
        eps = (z - self.mu.unsqueeze(0)) / std.unsqueeze(0)
        logp = -0.5 * ((eps**2) + 2*self.logstd.unsqueeze(0) + math.log(2*math.pi))
        return logp.sum(dim=-1)

policy = LatentGaussianPolicy(LATENT_DIM).to(DEVICE)
opt_policy = optim.Adam(policy.parameters(), lr=POLICY_LR)

# -------------------------
# VAE training (optional pretrain)
# -------------------------
if PRETRAINED_VAE and os.path.exists(PRETRAINED_VAE_PATH):
    print("Loading pretrained VAE:", PRETRAINED_VAE_PATH)
    vae.load_state_dict(torch.load(PRETRAINED_VAE_PATH, map_location=DEVICE))
else:
    print("Training GraphVAE (pretraining)...")
    bce = nn.BCELoss()
    ce = nn.CrossEntropyLoss(ignore_index=NODE_FEATURE_DIM-1)
    for ep in range(1, VAE_EPOCHS+1):
        vae.train()
        losses = []
        for A_b, X_b in batch_iter(A_array, X_array, batch_size=BATCH_SIZE):
            opt_vae.zero_grad()
            adj_pred, node_logits, mu, logvar = vae(A_b, X_b)
            # adj loss
            adj_loss = bce(adj_pred.view(-1), A_b.view(-1))
            # node loss: X_b is one-hot
            target = torch.argmax(X_b, dim=-1)
            node_loss = ce(node_logits.view(-1, node_logits.size(-1)), target.view(-1))
            # KL
            kld = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp()) / A_b.size(0)
            loss = adj_loss + node_loss + 0.1 * kld
            loss.backward()
            torch.nn.utils.clip_grad_norm_(vae.parameters(), 1.0)
            opt_vae.step()
            losses.append(loss.item())
        print(f"VAE Ep {ep}/{VAE_EPOCHS} loss={np.mean(losses):.4f}")
    torch.save(vae.state_dict(), os.path.join(RUN_DIR, "graphvae_pretrained.pt"))
    print("Saved pretrained VAE to", os.path.join(RUN_DIR, "graphvae_pretrained.pt"))

# -------------------------
# Reward function (QED) and helpers
# -------------------------
def reward_qed(smiles: str) -> float:
    if not smiles: return 0.0
    try:
        m = Chem.MolFromSmiles(smiles)
        if m is None: return 0.0
        return float(QED.qed(m))
    except Exception:
        return 0.0

def canonicalize_valid(smiles_list: List[str]) -> List[str]:
    out = []
    for s in smiles_list:
        if not s: continue
        try:
            m = Chem.MolFromSmiles(s)
            if m:
                out.append(Chem.MolToSmiles(m, canonical=True))
        except Exception:
            pass
    return out

# -------------------------
# PPO loop (policy-only): sample z, decode, compute rewards, update policy
# -------------------------
print("Starting PPO optimization of latent policy...")
running_baseline = None

all_generated = []  # accumulate final generated SMILES across PPO iterations

for it in range(1, PPO_EPOCHS+1):
    # sample a batch of latent vectors from policy
    z_batch, logp_old = policy.sample(PPO_BATCH)              # z_batch: (B, latent), logp_old: (B,)
    z_batch = z_batch.to(DEVICE)
    logp_old = logp_old.to(DEVICE)

    # decode each z into a SMILES (stochastically — sample atoms/edges)
    decoded_smiles = []
    for i_z in range(z_batch.size(0)):
        z = z_batch[i_z:i_z+1]   # shape (1, latent)
        adj_prob, node_logits = vae.decode_latent(z)   # adj_prob: (1,N,N), node_logits: (1,N,C)
        adj_np = adj_prob[0].cpu().numpy()
        node_prob = torch.softmax(node_logits[0], dim=-1).cpu().numpy()

        # try multiple temp settings and pick the best reward among them (optional)
        best_smi = ""
        best_r = -1.0
        for node_temp, adj_temp in SAMPLE_TEMPS:
            smi = graph_to_mol_stochastic(adj_np, node_prob, adj_temp=adj_temp, node_temp=node_temp)
            r = reward_qed(smi)
            if r > best_r:
                best_r = r; best_smi = smi
        decoded_smiles.append(best_smi)

    # compute rewards
    rewards = torch.tensor([reward_qed(s) for s in decoded_smiles], dtype=torch.float32, device=DEVICE)  # (B,)
    # baseline (running mean) for advantage
    batch_mean = rewards.mean().item()
    if running_baseline is None:
        running_baseline = batch_mean
    else:
        running_baseline = BASELINE_MOMENTUM * running_baseline + (1.0 - BASELINE_MOMENTUM) * batch_mean
    advantages = rewards - running_baseline
    advantages = (advantages - advantages.mean()) / (advantages.std() + 1e-8)

    # compute new log probs for sampled z under current policy
    logp_new = policy.log_prob(z_batch)

    # PPO clipped objective (we maximize, so minimize negative)
    ratio = torch.exp(logp_new - logp_old.detach())
    surr1 = ratio * advantages
    surr2 = torch.clamp(ratio, 1.0 - CLIP_EPS, 1.0 + CLIP_EPS) * advantages
    policy_loss = -torch.min(surr1, surr2).mean()

    # entropy bonus to encourage exploration
    std = torch.exp(policy.logstd)
    entropy = (0.5 * (1.0 + math.log(2 * math.pi)) + policy.logstd).sum()   # closed form for diagonal gaussian per-dim; summed dims
    policy_loss = policy_loss - ENT_COEF * entropy

    opt_policy.zero_grad()
    policy_loss.backward()
    torch.nn.utils.clip_grad_norm_(policy.parameters(), 1.0)
    opt_policy.step()

    # optionally collect samples for final output
    if it % 5 == 0 or it == PPO_EPOCHS:
        # sample some from mu (deterministic) and decode with multiple temps -> collect canonical SMILES
        z_det = policy.mu.unsqueeze(0)  # (1, latent)
        adj_prob, node_logits = vae.decode_latent(z_det.to(DEVICE))
        adj_np = adj_prob[0].cpu().numpy(); node_prob = torch.softmax(node_logits[0], dim=-1).cpu().numpy()
        # produce a handful of samples at different temperatures
        for node_temp, adj_temp in SAMPLE_TEMPS:
            for _ in range(20):
                smi = graph_to_mol_stochastic(adj_np, node_prob, adj_temp=adj_temp, node_temp=node_temp)
                if smi:
                    all_generated.append(smi)

    if it % 10 == 0:
        avg_r = rewards.mean().item()
        print(f"PPO it {it}/{PPO_EPOCHS}  avg_reward={avg_r:.4f}  baseline={running_baseline:.4f}  loss={policy_loss.item():.4f}")

# -------------------------
# Save combined generated SMILES and CSV
# -------------------------
# also add some purely random samples from policy to expand list
z_rand, _ = policy.sample(500)
z_rand = z_rand.to(DEVICE)
for i in range(z_rand.size(0)):
    adj_prob, node_logits = vae.decode_latent(z_rand[i:i+1])
    adj_np = adj_prob[0].cpu().numpy(); node_prob = torch.softmax(node_logits[0], dim=-1).cpu().numpy()
    smi = graph_to_mol_stochastic(adj_np, node_prob, adj_temp=1.0, node_temp=1.0)
    if smi:
        all_generated.append(smi)

# canonicalize & save unique first-occurrence
canonical = canonicalize_valid(all_generated)
unique = []
seen = set()
for s in canonical:
    if s not in seen:
        seen.add(s); unique.append(s)

with open(OUT_SMI, "w") as fh:
    for s in unique:
        fh.write(s + "\n")
with open(OUT_CSV, "w", newline="") as fh:
    w = csv.writer(fh); w.writerow(["index", "smiles"])
    for i,s in enumerate(unique): w.writerow([i, s])

print(f"Saved final generated SMILES to: {OUT_SMI} and CSV to: {OUT_CSV}")

# -------------------------
# Evaluate and save metrics
# -------------------------
def mean_pairwise_tanimoto(smiles_list, fp_radius=2, nbits=2048, sample_limit=2000):
    fps=[]
    for s in smiles_list[:sample_limit]:
        m = Chem.MolFromSmiles(s)
        if m:
            try:
                fps.append(AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=nbits))
            except:
                pass
    n=len(fps)
    if n<2: return 0.0
    s=0.0; cnt=0
    for i in range(n):
        for j in range(i+1,n):
            s+= DataStructs.TanimotoSimilarity(fps[i], fps[j]); cnt+=1
    return 1.0 - (s/cnt if cnt>0 else 0.0)

valid_smiles = canonical
metrics = {
    "n_generated": len(all_generated),
    "n_valid_unique": len(unique),
    "validity_fraction": len(valid_smiles)/max(1, len(all_generated)),
    "uniqueness_fraction": len(unique)/max(1, len(all_generated)),
    "avg_qed": float(np.mean([reward_qed(s) for s in valid_smiles])) if valid_smiles else 0.0,
    "mean_tanimoto_distance": float(mean_pairwise_tanimoto(valid_smiles))
}
with open(METRICS_JSON, "w") as fh:
    json.dump(metrics, fh, indent=2)
print("Saved metrics to", METRICS_JSON)
print("Metrics:", json.dumps(metrics, indent=2))
```

```python
# ============================================================
# GraphVAE + PPO with Safe Decoder (QED optimization)
# ============================================================

import os, json, csv, math
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from rdkit import Chem
from rdkit.Chem import QED, AllChem

# -------------------------
# Config
# -------------------------
DATA_SMI = "/kaggle/input/chembl-smi/chembl_molecules.smi"
RUN_DIR = "/kaggle/working/graphvae_main"
os.makedirs(RUN_DIR, exist_ok=True)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)

# GraphVAE hyperparams
MAX_ATOMS = 38
ATOM_LIST = ["C","N","O","S","F","Cl","Br","I","P","B","H"]
NODE_FEATURE_DIM = len(ATOM_LIST) + 1
HID_DIM = 128
LATENT_DIM = 64
BATCH_SIZE = 64
VAE_EPOCHS = 100
LR_VAE = 1e-3

# PPO hyperparams
PPO_EPOCHS = 500
PPO_BATCH = 128
POLICY_LR = 3e-4
CLIP_EPS = 0.2
ENT_COEF = 1e-3
BASELINE_MOMENTUM = 0.9

# output paths
OUT_SMI = os.path.join(RUN_DIR, "graphvae_ppo_generated.smi")
OUT_CSV = os.path.join(RUN_DIR, "graphvae_ppo_generated.csv")
METRICS_JSON = os.path.join(RUN_DIR, "graphvae_ppo_metrics.json")

# -------------------------
# Utils: SMILES <-> graph
# -------------------------
def atom_onehot(sym):
    if sym in ATOM_LIST: idx = ATOM_LIST.index(sym)
    else: idx = len(ATOM_LIST)
    v = np.zeros(NODE_FEATURE_DIM, dtype=np.float32); v[idx] = 1.0
    return v

def mol_to_graph(smi, max_atoms=MAX_ATOMS):
    try:
        m = Chem.MolFromSmiles(smi)
        if not m: return None
        n = m.GetNumAtoms()
        if n == 0 or n > max_atoms: return None
        A = np.zeros((max_atoms,max_atoms),dtype=np.float32)
        X = np.zeros((max_atoms,NODE_FEATURE_DIM),dtype=np.float32)
        for i,a in enumerate(m.GetAtoms()):
            X[i] = atom_onehot(a.GetSymbol())
        for b in m.GetBonds():
            i,j = b.GetBeginAtomIdx(), b.GetEndAtomIdx()
            A[i,j] = A[j,i] = 1.0
        return A,X
    except: return None

def load_smiles(path, limit=None):
    smis=[]
    with open(path) as f:
        for i,ln in enumerate(f):
            if not ln.strip(): continue
            smis.append(ln.strip().split()[0])
            if limit and i>=limit: break
    return smis

# -------------------------
# Safe decoder
# -------------------------
_ALLOWED_VALENCE = {"C":4,"N":3,"O":2,"S":2,"F":1,"Cl":1,"Br":1,"I":1,"P":3,"B":3,"H":1}

def safe_graph_to_mol_stochastic(A_prob, X_prob,
                                 adj_temp=1.0, node_temp=1.0,
                                 max_attempts=5):
    try:
        N = int(A_prob.shape[0])
        node_logits = np.log(np.clip(X_prob,1e-12,1.0))
        if node_temp != 1.0:
            node_logits = node_logits/float(node_temp)
        node_exp = np.exp(node_logits-node_logits.max(1,keepdims=True))
        node_probs = node_exp/(node_exp.sum(1,keepdims=True)+1e-12)
        top_idx = np.argmax(node_probs,axis=1)
        top_conf = np.max(node_probs,axis=1)
        active_mask = (top_idx < len(ATOM_LIST)) & (top_conf>0.2)

        A = np.clip(A_prob,1e-6,1-1e-6)
        if adj_temp!=1.0:
            logit = np.log(A/(1.0-A)); logit=logit/float(adj_temp)
            A = 1.0/(1.0+np.exp(-logit))

        for attempt in range(max_attempts):
            rw = Chem.RWMol(); idx_map={}
            for i in range(N):
                if not active_mask[i]: continue
                p=node_probs[i]; p=p/p.sum()
                ai=np.random.choice(len(p),p=p)
                if ai>=len(ATOM_LIST): continue
                sym=ATOM_LIST[ai]
                idx_map[i]=rw.AddAtom(Chem.Atom(sym))
            if len(idx_map)<1: continue
            current_degree={i:0 for i in idx_map}
            edges=[(float(A[i,j]),i,j) for i in idx_map for j in idx_map if j>i]
            edges.sort(reverse=True)
            for p,i,j in edges:
                if np.random.rand()>=p: continue
                si=rw.GetAtomWithIdx(idx_map[i]).GetSymbol()
                sj=rw.GetAtomWithIdx(idx_map[j]).GetSymbol()
                if current_degree[i]<_ALLOWED_VALENCE.get(si,4) and current_degree[j]<_ALLOWED_VALENCE.get(sj,4):
                    try:
                        rw.AddBond(idx_map[i],idx_map[j],Chem.BondType.SINGLE)
                        current_degree[i]+=1; current_degree[j]+=1
                    except: pass
            try:
                m=rw.GetMol(); Chem.SanitizeMol(m,catchErrors=False)
                smi=Chem.MolToSmiles(m,canonical=True)
                if smi: return smi
            except: continue
        return ""
    except: return ""

# -------------------------
# GraphVAE
# -------------------------
class Encoder(nn.Module):
    def __init__(self,in_dim,hid_dim,lat_dim):
        super().__init__()
        self.lin1=nn.Linear(in_dim,hid_dim)
        self.lin2=nn.Linear(hid_dim,hid_dim)
        self.out=nn.Linear(hid_dim,lat_dim*2)
    def forward(self,A,X):
        h=torch.matmul(A,X); h=torch.relu(self.lin1(h))
        h=torch.matmul(A,h); h=torch.relu(self.lin2(h))
        h_pool=h.mean(1); mu,logvar=self.out(h_pool).chunk(2,-1)
        return mu,logvar

class Decoder(nn.Module):
    def __init__(self,lat_dim,hid_dim,node_dim):
        super().__init__()
        self.node_proj=nn.Linear(lat_dim,hid_dim)
        self.node_mlp=nn.Sequential(nn.Linear(hid_dim,hid_dim),nn.ReLU(),
                                    nn.Linear(hid_dim,node_dim))
        self.score_proj=nn.Linear(hid_dim,hid_dim)
    def forward(self,z,max_nodes):
        B=z.size(0)
        node_lat=self.node_proj(z).unsqueeze(1).repeat(1,max_nodes,1)
        node_logits=self.node_mlp(node_lat)
        score=torch.matmul(self.score_proj(node_lat),self.score_proj(node_lat).transpose(1,2))
        adj_prob=torch.sigmoid(score)
        return adj_prob,node_logits

class GraphVAE(nn.Module):
    def __init__(self,node_dim,hid_dim,lat_dim,max_nodes):
        super().__init__()
        self.encoder=Encoder(node_dim,hid_dim,lat_dim)
        self.decoder=Decoder(lat_dim,hid_dim,node_dim)
        self.latent_dim=lat_dim; self.max_nodes=max_nodes
    def reparam(self,mu,logvar):
        std=torch.exp(0.5*logvar); eps=torch.randn_like(std)
        return mu+eps*std
    def forward(self,A,X):
        mu,logvar=self.encoder(A,X); z=self.reparam(mu,logvar)
        adj,node_logits=self.decoder(z,self.max_nodes)
        return adj,node_logits,mu,logvar
    def decode_latent(self,z):
        with torch.no_grad(): return self.decoder(z,self.max_nodes)

# -------------------------
# Data prep
# -------------------------
print("Loading dataset...")
smiles_all=load_smiles(DATA_SMI,limit=20000)
graphs=[mol_to_graph(s) for s in tqdm(smiles_all,desc="graphs")]
graphs=[g for g in graphs if g]
A_array=np.stack([g[0] for g in graphs])
X_array=np.stack([g[1] for g in graphs])

def batch_iter(A,X,batch=BATCH_SIZE,shuffle=True):
    idxs=np.arange(len(A)); 
    if shuffle: np.random.shuffle(idxs)
    for i in range(0,len(A),batch):
        b=idxs[i:i+batch]
        yield torch.tensor(A[b],dtype=torch.float32,device=DEVICE), \
              torch.tensor(X[b],dtype=torch.float32,device=DEVICE)

# -------------------------
# Train VAE
# -------------------------
vae=GraphVAE(NODE_FEATURE_DIM,HID_DIM,LATENT_DIM,MAX_ATOMS).to(DEVICE)
opt_vae=optim.Adam(vae.parameters(),lr=LR_VAE)
bce=nn.BCELoss(); ce=nn.CrossEntropyLoss(ignore_index=NODE_FEATURE_DIM-1)

for ep in range(1,VAE_EPOCHS+1):
    vae.train(); losses=[]
    for A_b,X_b in batch_iter(A_array,X_array):
        opt_vae.zero_grad()
        adj,node_logits,mu,logvar=vae(A_b,X_b)
        adj_loss=bce(adj.view(-1),A_b.view(-1))
        target=torch.argmax(X_b,-1)
        node_loss=ce(node_logits.view(-1,node_logits.size(-1)),target.view(-1))
        kld=-0.5*torch.sum(1+logvar-mu.pow(2)-logvar.exp())/A_b.size(0)
        loss=adj_loss+node_loss+0.1*kld
        loss.backward(); torch.nn.utils.clip_grad_norm_(vae.parameters(),1.0)
        opt_vae.step(); losses.append(loss.item())
    print(f"VAE Ep {ep} loss={np.mean(losses):.4f}")

# -------------------------
# PPO Policy
# -------------------------
class LatentGaussianPolicy(nn.Module):
    def __init__(self,latent_dim):
        super().__init__()
        self.mu=nn.Parameter(torch.zeros(latent_dim))
        self.logstd=nn.Parameter(torch.zeros(latent_dim))
    def sample(self,n):
        std=torch.exp(self.logstd)
        eps=torch.randn((n,self.mu.shape[0]),device=self.mu.device)
        z=self.mu.unsqueeze(0)+eps*std.unsqueeze(0)
        var=(std**2).unsqueeze(0)
        logp=-0.5*((eps**2)+2*self.logstd.unsqueeze(0)+math.log(2*math.pi))
        return z,logp.sum(-1)
    def log_prob(self,z):
        std=torch.exp(self.logstd)
        eps=(z-self.mu.unsqueeze(0))/std.unsqueeze(0)
        logp=-0.5*((eps**2)+2*self.logstd.unsqueeze(0)+math.log(2*math.pi))
        return logp.sum(-1)

policy=LatentGaussianPolicy(LATENT_DIM).to(DEVICE)
opt_policy=optim.Adam(policy.parameters(),lr=POLICY_LR)

def reward_qed(smi): 
    try: m=Chem.MolFromSmiles(smi); return float(QED.qed(m)) if m else 0.0
    except: return 0.0

# -------------------------
# PPO training
# -------------------------
print("Starting PPO...")
baseline=None; all_generated=[]
for it in range(1,PPO_EPOCHS+1):
    z_batch,logp_old=policy.sample(PPO_BATCH)
    z_batch=z_batch.to(DEVICE); logp_old=logp_old.to(DEVICE)
    decoded=[]
    for i in range(z_batch.size(0)):
        adj,node_logits=vae.decode_latent(z_batch[i:i+1])
        adj_np=adj[0].detach().cpu().numpy()
        node_np=torch.softmax(node_logits[0],-1).detach().cpu().numpy()
        smi=safe_graph_to_mol_stochastic(adj_np,node_np)
        decoded.append(smi)
    rewards=torch.tensor([reward_qed(s) for s in decoded],dtype=torch.float32,device=DEVICE)
    mean_r=rewards.mean().item()
    if baseline is None: baseline=mean_r
    else: baseline=BASELINE_MOMENTUM*baseline+(1-BASELINE_MOMENTUM)*mean_r
    adv=rewards-baseline
    adv=(adv-adv.mean())/(adv.std()+1e-8)
    logp_new=policy.log_prob(z_batch)
    ratio=torch.exp(logp_new-logp_old.detach())
    surr1=ratio*adv; surr2=torch.clamp(ratio,1-CLIP_EPS,1+CLIP_EPS)*adv
    loss=-torch.min(surr1,surr2).mean()
    std=torch.exp(policy.logstd)
    entropy=(0.5*(1+math.log(2*math.pi))+policy.logstd).sum()
    loss=loss-ENT_COEF*entropy
    opt_policy.zero_grad(); loss.backward()
    torch.nn.utils.clip_grad_norm_(policy.parameters(),1.0)
    opt_policy.step()
    if it%10==0: print(f"PPO {it}/{PPO_EPOCHS} avg_r={mean_r:.3f}")
    if it%5==0: all_generated+=decoded

# -------------------------
# Save outputs
# -------------------------
def canonicalize(smiles):
    out=[]; seen=set()
    for s in smiles:
        if not s: continue
        try:
            m=Chem.MolFromSmiles(s)
            if m:
                can=Chem.MolToSmiles(m,canonical=True)
                if can not in seen:
                    seen.add(can); out.append(can)
        except: pass
    return out

unique=canonicalize(all_generated)
with open(OUT_SMI,"w") as f: [f.write(s+"\n") for s in unique]
with open(OUT_CSV,"w",newline="") as f:
    w=csv.writer(f); w.writerow(["index","smiles"])
    [w.writerow([i,s]) for i,s in enumerate(unique)]

qeds=[reward_qed(s) for s in unique]
metrics={
    "n_generated":len(all_generated),
    "n_valid_unique":len(unique),
    "validity_fraction":len(unique)/max(1,len(all_generated)),
    "uniqueness_fraction":len(unique)/max(1,len(all_generated)),
    "avg_qed":float(np.mean(qeds)) if qeds else 0.0
}
with open(METRICS_JSON,"w") as f: json.dump(metrics,f,indent=2)
print("Saved metrics:",metrics)
```

```python
# ============================================================
# GraphVAE PPO – Full Evaluation Suite
# ============================================================
import os, csv, json
import numpy as np
from rdkit import Chem, DataStructs
from rdkit.Chem import QED, Crippen, Descriptors, rdMolDescriptors, AllChem
from rdkit.Chem.Scaffolds import MurckoScaffold

# -------------------------
# Config
# -------------------------
RUN_DIR = "/kaggle/working/graphvae_main"
GEN_SMI = os.path.join(RUN_DIR, "graphvae_ppo_generated.smi")
EVAL_CSV = os.path.join(RUN_DIR, "graphvae_ppo_eval.csv")
EVAL_JSON = os.path.join(RUN_DIR, "graphvae_ppo_eval_metrics.json")

# -------------------------
# Helpers
# -------------------------
def scaffold_bemis_murcko(smiles):
    try:
        m = Chem.MolFromSmiles(smiles)
        if not m: return ""
        scaf = MurckoScaffold.GetScaffoldForMol(m)
        return Chem.MolToSmiles(scaf, canonical=True) if scaf else ""
    except: return ""

def mean_pairwise_tanimoto(smiles_list, fp_radius=2, nbits=2048, sample_limit=2000):
    fps=[]
    for s in smiles_list[:sample_limit]:
        m = Chem.MolFromSmiles(s)
        if m:
            try:
                fps.append(AllChem.GetMorganFingerprintAsBitVect(m, fp_radius, nBits=nbits))
            except: pass
    n=len(fps)
    if n<2: return 0.0
    sim_sum, cnt = 0.0, 0
    for i in range(n):
        for j in range(i+1, n):
            sim_sum += DataStructs.TanimotoSimilarity(fps[i], fps[j])
            cnt += 1
    return 1.0 - (sim_sum/cnt if cnt>0 else 0.0)

# -------------------------
# Load generated SMILES
# -------------------------
smiles = []
with open(GEN_SMI,"r") as fh:
    for line in fh:
        s=line.strip()
        if s: smiles.append(s)

print(f"Loaded {len(smiles)} generated molecules.")

# -------------------------
# Evaluate each molecule
# -------------------------
records = []
valid = []
for smi in smiles:
    m = Chem.MolFromSmiles(smi)
    if not m: continue
    try:
        qed = QED.qed(m)
        mw = Descriptors.MolWt(m)
        logp = Crippen.MolLogP(m)
        tpsa = rdMolDescriptors.CalcTPSA(m)
        scaf = scaffold_bemis_murcko(smi)
        records.append({
            "smiles": smi,
            "qed": qed,
            "MolWt": mw,
            "LogP": logp,
            "TPSA": tpsa,
            "scaffold": scaf
        })
        valid.append(smi)
    except Exception:
        continue

# -------------------------
# Summary metrics
# -------------------------
valid_unique = list(dict.fromkeys(valid))
n_generated = len(smiles)
n_valid = len(valid)
n_unique = len(valid_unique)

metrics = {
    "n_generated": n_generated,
    "n_valid": n_valid,
    "validity_fraction": n_valid / max(1,n_generated),
    "n_unique": n_unique,
    "uniqueness_fraction": n_unique / max(1,n_valid),
    "avg_qed": float(np.mean([r["qed"] for r in records])) if records else 0.0,
    "MolWt_mean": float(np.mean([r["MolWt"] for r in records])) if records else 0.0,
    "LogP_mean": float(np.mean([r["LogP"] for r in records])) if records else 0.0,
    "TPSA_mean": float(np.mean([r["TPSA"] for r in records])) if records else 0.0,
    "scaffold_diversity_fraction": len(set(r["scaffold"] for r in records if r["scaffold"])) / max(1,n_valid),
    "mean_tanimoto_distance": mean_pairwise_tanimoto(valid_unique)
}

# -------------------------
# Save outputs
# -------------------------
with open(EVAL_CSV,"w",newline="") as fh:
    w=csv.DictWriter(fh,fieldnames=list(records[0].keys()))
    w.writeheader()
    w.writerows(records)

with open(EVAL_JSON,"w") as fh:
    json.dump(metrics, fh, indent=2)

print("Saved per-molecule CSV to:", EVAL_CSV)
print("Saved metrics JSON to:", EVAL_JSON)
print("Metrics:", json.dumps(metrics, indent=2))
```
